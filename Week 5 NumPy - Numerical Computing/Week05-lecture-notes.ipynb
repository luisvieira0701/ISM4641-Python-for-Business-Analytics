## Introduction: The Foundation of Data Science in Python

Imagine you're a business analyst at a Fortune 500 company, and you need to analyze sales data for 10,000 products across 50 stores over 365 days. That's 182.5 million data points! Using Python's built-in lists would be painfully slow and memory-intensive. This is where **NumPy** (Numerical Python) becomes essential.

NumPy is the foundational library for numerical computing in Python. It provides:

- **High-performance multidimensional arrays** for storing and manipulating numerical data
- **Vectorized operations** that operate on entire arrays at once
- **Broadcasting** capabilities for operations on arrays of different shapes
- **Comprehensive mathematical functions** for statistical analysis

### Why NumPy Matters for Business Analytics

| Scenario | Python Lists | NumPy Arrays |
|----------|-------------|--------------|
| Add 1 million numbers | ~200 ms | ~2 ms (100x faster) |
| Store 1 million integers | ~28 MB | ~4 MB (7x smaller) |
| Apply discount to all prices | Loop required | Single operation |
| Calculate statistics | Manual functions | Built-in methods |

NumPy is the foundation of the entire Python data science ecosystem. Libraries like Pandas (data manipulation), Matplotlib (visualization), and Scikit-learn (machine learning) all build on NumPy arrays.

In this chapter, we'll master NumPy to prepare you for real-world business analytics challenges.

## 5.1 Getting Started with NumPy

NumPy is not part of Python's standard library, but it comes pre-installed in most data science environments like Google Colab and Anaconda. The convention is to import it with the alias `np`:

```python
import numpy as np

# Verify the import and check version
print(f"NumPy version: {np.__version__}")
print("NumPy is ready to use!")
```

### The ndarray: NumPy's Core Data Structure

The fundamental object in NumPy is the `ndarray` (n-dimensional array). Unlike Python lists, NumPy arrays:

- Store elements of a **single data type** (homogeneous)
- Are stored in **contiguous memory** (making them fast)
- Support **vectorized operations** (operate on all elements at once)
- Have a **fixed size** at creation (though you can create new arrays)

```python
# Create a simple NumPy array from a Python list
sales = np.array([125.50, 89.99, 245.00, 156.75, 310.25])

print("Sales array:", sales)
print(f"Type: {type(sales)}")
print(f"Data type of elements: {sales.dtype}")
print(f"Number of elements: {sales.size}")
print(f"Shape: {sales.shape}")
print(f"Number of dimensions: {sales.ndim}")
```

### Array Attributes Explained

| Attribute | Description | Example |
|-----------|-------------|--------|
| `dtype` | Data type of elements | `float64`, `int32`, `bool` |
| `shape` | Dimensions of array | `(5,)` for 1D, `(3, 4)` for 2D |
| `size` | Total number of elements | 5, 12 |
| `ndim` | Number of dimensions | 1, 2, 3 |
| `itemsize` | Size of each element in bytes | 8 for float64 |
| `nbytes` | Total memory used | size × itemsize |

```python
# Compare Python list vs NumPy array
import sys

# Python list
py_list = list(range(1000))

# NumPy array
np_array = np.arange(1000)

print("Memory Comparison:")
print(f"Python list: {sys.getsizeof(py_list)} bytes")
print(f"NumPy array: {np_array.nbytes} bytes")
print(f"NumPy uses {sys.getsizeof(py_list) / np_array.nbytes:.1f}x less memory!")
```

## 5.2 Creating Arrays

NumPy provides many ways to create arrays. Let's explore the most common methods.

### Creating Arrays from Python Lists

```python
# 1D array from a list
prices = np.array([29.99, 49.99, 79.99, 99.99, 149.99])
print("1D array (prices):")
print(prices)
print(f"Shape: {prices.shape}\n")

# 2D array from nested lists
quarterly_sales = np.array([
    [100, 120, 140, 130],  # Product A - Q1, Q2, Q3, Q4
    [80, 95, 110, 105],    # Product B
    [200, 220, 250, 240]   # Product C
])
print("2D array (quarterly sales):")
print(quarterly_sales)
print(f"Shape: {quarterly_sales.shape} (3 products, 4 quarters)")
print(f"Dimensions: {quarterly_sales.ndim}")
```

```python
# 3D array - e.g., sales by store, product, and quarter
store_sales = np.array([
    # Store 1
    [[100, 110, 120, 115],   # Product A
     [80, 85, 90, 88]],      # Product B
    # Store 2
    [[150, 160, 170, 165],   # Product A
     [120, 125, 130, 128]]   # Product B
])

print("3D array (store × product × quarter):")
print(store_sales)
print(f"Shape: {store_sales.shape}")
print(f"Dimensions: {store_sales.ndim}")
```

### Specifying Data Types

NumPy supports various data types. You can specify the type when creating arrays:

```python
# Different data types
integers = np.array([1, 2, 3, 4, 5], dtype=np.int32)
floats = np.array([1, 2, 3, 4, 5], dtype=np.float64)
booleans = np.array([True, False, True, False])

print(f"Integers: {integers}, dtype: {integers.dtype}")
print(f"Floats: {floats}, dtype: {floats.dtype}")
print(f"Booleans: {booleans}, dtype: {booleans.dtype}")
```

```python
# Common NumPy data types
print("Common NumPy Data Types:")
print("="*50)
data_types = [
    ("int32", "32-bit integer (-2B to 2B)"),
    ("int64", "64-bit integer (very large range)"),
    ("float32", "32-bit floating point"),
    ("float64", "64-bit floating point (default)"),
    ("bool", "Boolean (True/False)"),
    ("str", "String/Unicode")
]

for dtype, description in data_types:
    print(f"{dtype:12} {description}")
```

### Array Generation Functions

NumPy provides convenient functions to create common array patterns:

```python
# Arrays of zeros - useful for initialization
zeros_1d = np.zeros(5)
zeros_2d = np.zeros((3, 4))  # 3 rows, 4 columns

print("Zeros (1D):", zeros_1d)
print("\nZeros (3x4):")
print(zeros_2d)
```

```python
# Arrays of ones
ones = np.ones((2, 3))
print("Ones (2x3):")
print(ones)

# Arrays filled with a specific value
default_price = np.full((3, 4), 9.99)
print("\nDefault prices (3x4):")
print(default_price)
```

```python
# Identity matrix (useful in linear algebra)
identity = np.eye(4)
print("4x4 Identity matrix:")
print(identity)
```

```python
# np.arange - like Python's range() but returns an array
# Syntax: np.arange(start, stop, step)

months = np.arange(1, 13)  # 1 to 12
print("Months:", months)

even_numbers = np.arange(0, 20, 2)  # 0, 2, 4, ..., 18
print("Even numbers:", even_numbers)

countdown = np.arange(10, 0, -1)  # 10, 9, 8, ..., 1
print("Countdown:", countdown)
```

```python
# np.linspace - evenly spaced numbers over a range
# Syntax: np.linspace(start, stop, num_points)

# 5 evenly spaced points from 0 to 100
percentages = np.linspace(0, 100, 5)
print("Percentages:", percentages)

# 11 points from 0 to 1 (useful for probabilities)
probabilities = np.linspace(0, 1, 11)
print("Probabilities:", probabilities)

# Key difference: arange uses step size, linspace uses number of points
print("\narange(0, 10, 2):", np.arange(0, 10, 2))  # [0 2 4 6 8]
print("linspace(0, 10, 5):", np.linspace(0, 10, 5))  # [0 2.5 5 7.5 10]
```

### Random Array Generation

NumPy's random module is essential for simulations, sampling, and testing:

```python
# Set seed for reproducibility
np.random.seed(42)

# Random integers
random_sales = np.random.randint(50, 200, size=10)  # 10 random integers between 50-199
print("Random sales:", random_sales)

# Random floats between 0 and 1
random_rates = np.random.random(5)
print("\nRandom rates (0-1):", random_rates)

# Random floats in a range
random_prices = np.random.uniform(10, 100, 5)  # Between 10 and 100
print("\nRandom prices ($10-$100):", np.round(random_prices, 2))
```

```python
# Normal (Gaussian) distribution - very common in business data
np.random.seed(42)

# Simulate daily sales with mean $1500, std $300
daily_sales = np.random.normal(loc=1500, scale=300, size=30)
daily_sales = np.round(daily_sales, 2)

print("Simulated daily sales (30 days):")
print(daily_sales)
print(f"\nMean: ${daily_sales.mean():.2f}")
print(f"Std Dev: ${daily_sales.std():.2f}")
```

```python
# 2D random array
np.random.seed(42)
random_matrix = np.random.randint(1, 100, size=(3, 4))
print("Random 3x4 matrix:")
print(random_matrix)
```

## 5.3 Array Operations: The Power of Vectorization

**Vectorization** is NumPy's killer feature. Instead of using loops to process each element, you operate on the entire array at once. This is not just more concise—it's dramatically faster.

### Arithmetic Operations

```python
# Element-wise operations between arrays
prices = np.array([29.99, 49.99, 79.99, 99.99, 149.99])
quantities = np.array([100, 75, 50, 30, 20])

# Calculate revenue for each product
revenue = prices * quantities
print("Prices:", prices)
print("Quantities:", quantities)
print("Revenue (price × qty):", revenue)
print(f"Total Revenue: ${revenue.sum():,.2f}")
```

```python
# Scalar operations (apply to all elements)
prices = np.array([100, 200, 300, 400, 500])

print("Original prices:", prices)
print("After 10% discount:", prices * 0.90)
print("Add $5 shipping:", prices + 5)
print("With 7% tax:", prices * 1.07)
print("Double the prices:", prices * 2)
print("Prices divided by 10:", prices / 10)
```

```python
# Comparison: Python list vs NumPy array for the same operation
import time

# Python list approach
py_prices = list(range(1000000))
start = time.time()
py_discounted = [p * 0.90 for p in py_prices]
py_time = time.time() - start

# NumPy array approach
np_prices = np.arange(1000000)
start = time.time()
np_discounted = np_prices * 0.90
np_time = time.time() - start

print(f"Python list time: {py_time*1000:.2f} ms")
print(f"NumPy array time: {np_time*1000:.2f} ms")
print(f"NumPy is {py_time/np_time:.1f}x faster!")
```

### All Arithmetic Operators

```python
a = np.array([10, 20, 30, 40, 50])
b = np.array([2, 4, 5, 8, 10])

print(f"a = {a}")
print(f"b = {b}")
print()
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
print(f"a // b (floor division) = {a // b}")
print(f"a % b (modulo) = {a % b}")
print(f"a ** 2 (power) = {a ** 2}")
```

### Mathematical Functions

NumPy provides a comprehensive library of mathematical functions that operate on entire arrays:

```python
values = np.array([1, 4, 9, 16, 25, 36])

print("Original values:", values)
print()
print("Square root:", np.sqrt(values))
print("Squared:", np.square(values))
print("Cube:", np.power(values, 3))
print("Exponential (e^x):", np.exp(values[:3]))  # First 3 to avoid huge numbers
print("Natural log:", np.log(values))
print("Log base 10:", np.log10(values))
print("Log base 2:", np.log2(values))
```

```python
# Rounding functions
prices = np.array([9.99, 19.50, 29.95, 39.01, 49.99, 59.49])

print("Original prices:", prices)
print("Floor (round down):", np.floor(prices))
print("Ceil (round up):", np.ceil(prices))
print("Round (nearest):", np.round(prices))
print("Round to 1 decimal:", np.round(prices, 1))
print("Truncate (remove decimal):", np.trunc(prices))
```

```python
# Absolute value and sign
profit_loss = np.array([1500, -800, 2300, -150, 900, -1200])

print("Profit/Loss:", profit_loss)
print("Absolute values:", np.abs(profit_loss))
print("Signs (-1, 0, or 1):", np.sign(profit_loss))
```

### Comparison Operations

Comparisons return boolean arrays, which are powerful for filtering:

```python
sales = np.array([125, 89, 245, 56, 310, 178, 95])
target = 150

print(f"Sales: {sales}")
print(f"Target: {target}")
print()
print(f"Above target (> {target}): {sales > target}")
print(f"At or above target (>= {target}): {sales >= target}")
print(f"Below target (< {target}): {sales < target}")
print(f"Equal to 89: {sales == 89}")
print(f"Not equal to 89: {sales != 89}")
```

```python
# Array comparison
this_year = np.array([100, 120, 140, 130])
last_year = np.array([95, 125, 130, 135])

print("This year:", this_year)
print("Last year:", last_year)
print("Improved this year:", this_year > last_year)
```

## 5.4 Indexing and Slicing

Accessing specific elements or subsets of arrays is fundamental to data analysis.

### 1D Array Indexing

```python
# Weekly sales data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sales = np.array([1250, 980, 1567, 890, 2100, 2450, 1780])

print("Weekly Sales:")
for i, (day, amount) in enumerate(zip(days, sales)):
    print(f"  {day}: ${amount}")

print("\nIndexing Examples:")
print(f"First day (index 0): {days[0]} = ${sales[0]}")
print(f"Last day (index -1): {days[-1]} = ${sales[-1]}")
print(f"Third day (index 2): {days[2]} = ${sales[2]}")
print(f"Second to last (index -2): {days[-2]} = ${sales[-2]}")
```

```python
# Slicing: array[start:stop:step]
sales = np.array([1250, 980, 1567, 890, 2100, 2450, 1780])

print("All sales:", sales)
print()
print("First 3 days ([:3]):", sales[:3])
print("Last 3 days ([-3:]):", sales[-3:])
print("Middle days ([2:5]):", sales[2:5])
print("Every other day ([::2]):", sales[::2])
print("Reversed ([::-1]):", sales[::-1])
print("Weekdays only ([:5]):", sales[:5])
print("Weekend only ([5:]):", sales[5:])
```

### 2D Array Indexing

For 2D arrays, use `array[row, column]` notation:

```python
# Sales by region (rows) and quarter (columns)
regions = ['North', 'South', 'East', 'West']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

quarterly_sales = np.array([
    [100, 120, 140, 130],  # North
    [80, 95, 110, 105],    # South
    [150, 170, 190, 180],  # East
    [90, 100, 115, 110]    # West
])

print("Quarterly Sales (in thousands):")
print(f"{'':8}", end='')
for q in quarters:
    print(f"{q:>6}", end='')
print()
for i, region in enumerate(regions):
    print(f"{region:8}", end='')
    for j in range(4):
        print(f"{quarterly_sales[i, j]:>6}", end='')
    print()
```

```python
print("\n2D Indexing Examples:")
print(f"North region (row 0): {quarterly_sales[0]}")
print(f"Q1 all regions (col 0): {quarterly_sales[:, 0]}")
print(f"North Q2 (row 0, col 1): {quarterly_sales[0, 1]}")
print(f"East Q4 (row 2, col 3): {quarterly_sales[2, 3]}")
```

```python
print("\n2D Slicing Examples:")
print("First two regions, first two quarters:")
print(quarterly_sales[:2, :2])

print("\nLast two regions, last two quarters:")
print(quarterly_sales[-2:, -2:])

print("\nAll regions, Q2 and Q3:")
print(quarterly_sales[:, 1:3])
```

### Boolean Indexing (Filtering)

Boolean indexing is one of NumPy's most powerful features for data analysis. You create a boolean mask and use it to select elements:

```python
sales = np.array([125, 89, 245, 56, 310, 178, 95, 420, 67])

# Step 1: Create a boolean mask
high_sales_mask = sales > 150
print("Sales:", sales)
print("Mask (> 150):", high_sales_mask)

# Step 2: Use mask to filter
high_sales = sales[high_sales_mask]
print("\nHigh sales (> $150):", high_sales)
```

```python
# One-liner filtering
sales = np.array([125, 89, 245, 56, 310, 178, 95, 420, 67])

print("Sales:", sales)
print("Above 200:", sales[sales > 200])
print("Below 100:", sales[sales < 100])
print("Equal to 178:", sales[sales == 178])
```

```python
# Combining conditions with & (and) and | (or)
sales = np.array([125, 89, 245, 56, 310, 178, 95, 420, 67])

# Sales between 100 and 200
mid_range = sales[(sales >= 100) & (sales <= 200)]
print("Sales between $100 and $200:", mid_range)

# Sales below 100 OR above 300
extreme = sales[(sales < 100) | (sales > 300)]
print("Extreme sales (<$100 or >$300):", extreme)

# NOT: sales that are NOT below 100
not_low = sales[~(sales < 100)]
print("Not low sales (NOT <$100):", not_low)
```

```python
# Using np.where() - returns indices where condition is True
sales = np.array([125, 89, 245, 56, 310, 178, 95, 420, 67])

# Find indices of high sales
high_indices = np.where(sales > 200)[0]
print("Indices of sales > $200:", high_indices)
print("Values at those indices:", sales[high_indices])
```

```python
# np.where() with replacement - conditional values
sales = np.array([125, 89, 245, 56, 310, 178, 95])

# Label as 'High' or 'Low' based on threshold
labels = np.where(sales >= 150, 'High', 'Low')
print("Sales:", sales)
print("Labels:", labels)

# Apply bonus: 10% for high performers, 5% for others
bonuses = np.where(sales >= 150, sales * 0.10, sales * 0.05)
print("\nBonuses:", bonuses)
```

### Fancy Indexing

Select multiple specific elements using a list of indices:

```python
products = np.array(['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'])
prices = np.array([999.99, 29.99, 79.99, 299.99, 89.99])

# Select specific indices
featured_indices = [0, 2, 4]  # Laptop, Keyboard, Webcam
print("Featured products:", products[featured_indices])
print("Featured prices:", prices[featured_indices])
```

## 5.5 Statistical Functions

NumPy provides comprehensive statistical functions essential for business analytics.

### Basic Descriptive Statistics

```python
# Monthly sales data
monthly_sales = np.array([12500, 9800, 15670, 8900, 21000, 14500, 
                          17800, 13200, 18900, 16500, 22000, 25000])

print("Monthly Sales Analysis")
print("="*40)
print(f"Data: {monthly_sales}")
print()
print("--- Central Tendency ---")
print(f"Mean: ${monthly_sales.mean():,.2f}")
print(f"Median: ${np.median(monthly_sales):,.2f}")
print()
print("--- Spread ---")
print(f"Minimum: ${monthly_sales.min():,.2f}")
print(f"Maximum: ${monthly_sales.max():,.2f}")
print(f"Range: ${monthly_sales.max() - monthly_sales.min():,.2f}")
print(f"Variance: {monthly_sales.var():,.2f}")
print(f"Standard Deviation: ${monthly_sales.std():,.2f}")
print()
print("--- Aggregates ---")
print(f"Count: {monthly_sales.size}")
print(f"Sum: ${monthly_sales.sum():,.2f}")
```

```python
# Percentiles and Quartiles
monthly_sales = np.array([12500, 9800, 15670, 8900, 21000, 14500, 
                          17800, 13200, 18900, 16500, 22000, 25000])

print("Percentile Analysis")
print("="*40)
print(f"Min (0th): ${np.percentile(monthly_sales, 0):,.2f}")
print(f"25th percentile (Q1): ${np.percentile(monthly_sales, 25):,.2f}")
print(f"50th percentile (Median): ${np.percentile(monthly_sales, 50):,.2f}")
print(f"75th percentile (Q3): ${np.percentile(monthly_sales, 75):,.2f}")
print(f"Max (100th): ${np.percentile(monthly_sales, 100):,.2f}")

# Interquartile Range (IQR)
q1 = np.percentile(monthly_sales, 25)
q3 = np.percentile(monthly_sales, 75)
iqr = q3 - q1
print(f"\nIQR (Q3 - Q1): ${iqr:,.2f}")
```

### Finding Positions: argmin and argmax

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales = np.array([12500, 9800, 15670, 8900, 21000, 14500, 
                          17800, 13200, 18900, 16500, 22000, 25000])

# Find index of min and max
best_month_idx = monthly_sales.argmax()
worst_month_idx = monthly_sales.argmin()

print(f"Best month: {months[best_month_idx]} (${monthly_sales[best_month_idx]:,})")
print(f"Worst month: {months[worst_month_idx]} (${monthly_sales[worst_month_idx]:,})")
```

### Cumulative Functions

```python
# Running totals and products
daily_profits = np.array([100, -50, 200, 150, -30, 180, 220])

print("Daily Profits:", daily_profits)
print("Cumulative Sum:", np.cumsum(daily_profits))
print(f"Final Total: ${np.cumsum(daily_profits)[-1]}")
```

```python
# Year-to-date (YTD) sales tracking
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
monthly_sales = np.array([12500, 9800, 15670, 8900, 21000, 14500])
ytd_sales = np.cumsum(monthly_sales)

print("Year-to-Date Sales Tracking")
print("="*45)
print(f"{'Month':<8} {'Monthly':>12} {'YTD':>12}")
print("-"*45)
for month, sales, ytd in zip(months, monthly_sales, ytd_sales):
    print(f"{month:<8} ${sales:>10,} ${ytd:>10,}")
```

### Statistics Along Axes

For 2D arrays, you can calculate statistics along rows (axis=1) or columns (axis=0):

- **axis=0**: Operate along rows (result has shape of columns)
- **axis=1**: Operate along columns (result has shape of rows)

```python
# Sales by region (rows) and quarter (columns)
regions = ['North', 'South', 'East', 'West']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

quarterly_sales = np.array([
    [100, 120, 140, 130],  # North
    [80, 95, 110, 105],    # South
    [150, 170, 190, 180],  # East
    [90, 100, 115, 110]    # West
])

print("Quarterly Sales (in thousands):")
print(quarterly_sales)

print("\n--- Axis Operations ---")
# Total by region (sum across columns for each row)
region_totals = quarterly_sales.sum(axis=1)
print(f"Total by region (axis=1): {region_totals}")

# Total by quarter (sum across rows for each column)
quarter_totals = quarterly_sales.sum(axis=0)
print(f"Total by quarter (axis=0): {quarter_totals}")

# Grand total
print(f"Grand total: {quarterly_sales.sum()}")
```

```python
# Complete regional analysis
print("\nRegional Performance Summary:")
print("="*60)
print(f"{'Region':<10} {'Total':>10} {'Mean':>10} {'Min':>8} {'Max':>8} {'Best Q'}")
print("-"*60)

for i, region in enumerate(regions):
    total = quarterly_sales[i].sum()
    mean = quarterly_sales[i].mean()
    minimum = quarterly_sales[i].min()
    maximum = quarterly_sales[i].max()
    best_q = quarters[quarterly_sales[i].argmax()]
    print(f"{region:<10} ${total:>8,} ${mean:>8.1f} ${minimum:>6,} ${maximum:>6,} {best_q:>6}")
```

```python
# Quarterly performance summary
print("\nQuarterly Performance Summary:")
print("="*50)
print(f"{'Quarter':<10} {'Total':>10} {'Mean':>10} {'Best Region'}")
print("-"*50)

for j, quarter in enumerate(quarters):
    total = quarterly_sales[:, j].sum()
    mean = quarterly_sales[:, j].mean()
    best_region = regions[quarterly_sales[:, j].argmax()]
    print(f"{quarter:<10} ${total:>8,} ${mean:>8.1f} {best_region:>12}")
```

## 5.6 Broadcasting

**Broadcasting** allows NumPy to perform operations on arrays of different shapes. This is incredibly powerful for business calculations.

### Scalar Broadcasting

```python
# Apply a scalar operation to entire array
prices = np.array([100, 200, 300, 400, 500])

# The scalar 1.07 is "broadcast" to match the array shape
prices_with_tax = prices * 1.07

print("Original prices:", prices)
print("With 7% tax:", prices_with_tax)
```

### Array Broadcasting

When operating on two arrays, NumPy broadcasts them to compatible shapes:

```python
# 2D array: 3 products × 4 quarters
quarterly_sales = np.array([
    [100, 120, 140, 130],  # Product A
    [80, 95, 110, 105],    # Product B
    [150, 170, 190, 180]   # Product C
])

# 1D array: discount rate per quarter
quarterly_discounts = np.array([0.95, 0.90, 0.85, 0.80])  # 5%, 10%, 15%, 20% off

# Broadcasting: 1D array is broadcast across each row
discounted_sales = quarterly_sales * quarterly_discounts

print("Original sales:")
print(quarterly_sales)
print("\nDiscount rates:", quarterly_discounts)
print("\nDiscounted sales:")
print(discounted_sales)
```

```python
# Different discount per product (column broadcasting)
quarterly_sales = np.array([
    [100, 120, 140, 130],  # Product A
    [80, 95, 110, 105],    # Product B
    [150, 170, 190, 180]   # Product C
])

# Discount rate per product (needs to be column vector)
product_discounts = np.array([[0.95], [0.90], [0.85]])  # Shape (3, 1)

# Or use reshape
product_discounts = np.array([0.95, 0.90, 0.85]).reshape(3, 1)

discounted = quarterly_sales * product_discounts

print("Original sales:")
print(quarterly_sales)
print("\nProduct discounts (column):")
print(product_discounts)
print("\nDiscounted by product:")
print(discounted)
```

```python
# Practical example: Calculate prices with different tax rates by state
# Rows = products, Columns = states
base_prices = np.array([
    [10.00],   # Product 1
    [25.00],   # Product 2
    [50.00]    # Product 3
])

state_tax_rates = np.array([1.06, 1.07, 1.08, 1.0725])  # FL, CA, NY, TX
states = ['Florida', 'California', 'New York', 'Texas']

# Broadcasting creates 3×4 matrix of prices with tax
prices_with_tax = base_prices * state_tax_rates

print("Prices with state tax:")
print(f"{'Product':<10}", end='')
for state in states:
    print(f"{state:>12}", end='')
print()
for i, row in enumerate(prices_with_tax):
    print(f"Product {i+1:<3}", end='')
    for price in row:
        print(f"${price:>10.2f}", end='')
    print()
```

### Broadcasting Rules

Two arrays are compatible for broadcasting when for each dimension:
1. The dimensions are equal, OR
2. One of them is 1

NumPy compares shapes element-wise from right to left.

```python
# Compatible shapes for broadcasting
print("Broadcasting Examples:")
print("="*40)

# (3,) with (3,) - same shape
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])
print(f"(3,) + (3,) = {a + b}")

# (3,) with scalar - scalar broadcasts
print(f"(3,) + scalar = {a + 10}")

# (3, 4) with (4,) - 1D broadcasts across rows
c = np.ones((3, 4))
d = np.array([1, 2, 3, 4])
print(f"(3,4) + (4,):")
print(c + d)

# (3, 4) with (3, 1) - column broadcasts across columns
e = np.array([[10], [20], [30]])
print(f"\n(3,4) + (3,1):")
print(c + e)
```

```python
# argsort - get indices that would sort the array
products = np.array(['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'])
sales = np.array([125, 89, 245, 56, 310])

# Get indices that sort by sales
sorted_indices = np.argsort(sales)[::-1]  # Descending

print("Products ranked by sales (highest first):")
for i, idx in enumerate(sorted_indices, 1):
    print(f"  {i}. {products[idx]}: ${sales[idx]}")
```

## 5.9 Handling Missing Data

Real-world data often contains missing values. NumPy uses `np.nan` (Not a Number) to represent missing numeric data.

```python
# Array with missing values
sales = np.array([125.0, np.nan, 245.0, 156.0, np.nan, 310.0, 278.0])

print("Sales with missing values:", sales)

# Standard functions return nan if any nan present
print(f"\nStandard mean: {sales.mean()}")  # Returns nan!
print(f"Standard sum: {sales.sum()}")  # Returns nan!
```

```python
# Use nan-aware functions
sales = np.array([125.0, np.nan, 245.0, 156.0, np.nan, 310.0, 278.0])

print("Using nan-aware functions:")
print(f"nanmean: ${np.nanmean(sales):.2f}")
print(f"nansum: ${np.nansum(sales):.2f}")
print(f"nanmin: ${np.nanmin(sales):.2f}")
print(f"nanmax: ${np.nanmax(sales):.2f}")
print(f"nanstd: ${np.nanstd(sales):.2f}")
```

```python
# Detecting and counting missing values
sales = np.array([125.0, np.nan, 245.0, 156.0, np.nan, 310.0, 278.0])

# Check for nan
nan_mask = np.isnan(sales)
print("NaN mask:", nan_mask)
print(f"Number of missing values: {np.sum(nan_mask)}")
print(f"Number of valid values: {np.sum(~nan_mask)}")

# Get only valid values
valid_sales = sales[~np.isnan(sales)]
print(f"Valid sales: {valid_sales}")
```

```python
# Replace missing values
sales = np.array([125.0, np.nan, 245.0, 156.0, np.nan, 310.0, 278.0])

# Replace with mean
mean_value = np.nanmean(sales)
sales_filled = np.where(np.isnan(sales), mean_value, sales)

print(f"Original: {sales}")
print(f"Mean: ${mean_value:.2f}")
print(f"Filled with mean: {sales_filled}")
```

### Example 2: Product Portfolio Analysis

```python
# Product data
products = np.array(['Laptop Pro', 'Wireless Mouse', 'Mech Keyboard', 
                     '4K Monitor', 'USB Hub', 'Webcam HD'])
prices = np.array([999.99, 49.99, 149.99, 399.99, 29.99, 79.99])
costs = np.array([650.00, 20.00, 75.00, 250.00, 12.00, 35.00])
units_sold = np.array([150, 500, 300, 200, 800, 400])

# Calculate metrics
revenue = prices * units_sold
total_cost = costs * units_sold
profit = revenue - total_cost
margin_pct = (profit / revenue) * 100

print("="*80)
print("                    PRODUCT PORTFOLIO ANALYSIS")
print("="*80)
print(f"{'Product':<16} {'Price':>10} {'Units':>8} {'Revenue':>12} {'Profit':>12} {'Margin':>8}")
print("-"*80)

for i in range(len(products)):
    print(f"{products[i]:<16} ${prices[i]:>8.2f} {units_sold[i]:>8} "
          f"${revenue[i]:>10,.2f} ${profit[i]:>10,.2f} {margin_pct[i]:>7.1f}%")

print("-"*80)
print(f"{'TOTALS':<16} {'':<10} {units_sold.sum():>8} ${revenue.sum():>10,.2f} ${profit.sum():>10,.2f}")

# Analysis
print("\n--- Key Insights ---")
best_margin_idx = margin_pct.argmax()
best_profit_idx = profit.argmax()
best_volume_idx = units_sold.argmax()

print(f"Highest Margin: {products[best_margin_idx]} ({margin_pct[best_margin_idx]:.1f}%)")
print(f"Most Profitable: {products[best_profit_idx]} (${profit[best_profit_idx]:,.2f})")
print(f"Best Seller: {products[best_volume_idx]} ({units_sold[best_volume_idx]} units)")
print(f"Average Margin: {margin_pct.mean():.1f}%")
```

### Example 3: Year-over-Year Regional Analysis

```python
# Quarterly data for 4 regions over 2 years
regions = ['North', 'South', 'East', 'West']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

# 2024 sales (in thousands)
sales_2024 = np.array([
    [100, 120, 140, 130],  # North
    [80, 95, 110, 105],    # South
    [150, 170, 190, 180],  # East
    [90, 100, 115, 110]    # West
])

# 2025 sales (in thousands)
sales_2025 = np.array([
    [115, 135, 155, 145],  # North
    [92, 108, 125, 118],   # South
    [168, 188, 210, 198],  # East
    [102, 115, 130, 125]   # West
])

# Calculate growth
growth_dollars = sales_2025 - sales_2024
growth_percent = ((sales_2025 - sales_2024) / sales_2024) * 100

print("="*70)
print("            YEAR-OVER-YEAR REGIONAL GROWTH ANALYSIS")
print("="*70)

# Regional summary
print("\n--- Regional Performance ---")
print(f"{'Region':<10} {'2024':>12} {'2025':>12} {'Growth $':>12} {'Growth %':>10}")
print("-"*70)

for i, region in enumerate(regions):
    y2024 = sales_2024[i].sum()
    y2025 = sales_2025[i].sum()
    growth_d = y2025 - y2024
    growth_p = (growth_d / y2024) * 100
    print(f"{region:<10} ${y2024:>10,}K ${y2025:>10,}K ${growth_d:>10,}K {growth_p:>9.1f}%")

# Quarterly summary
print("\n--- Quarterly Performance ---")
print(f"{'Quarter':<10} {'2024':>12} {'2025':>12} {'Growth $':>12} {'Growth %':>10}")
print("-"*70)

for j, quarter in enumerate(quarters):
    y2024 = sales_2024[:, j].sum()
    y2025 = sales_2025[:, j].sum()
    growth_d = y2025 - y2024
    growth_p = (growth_d / y2024) * 100
    print(f"{quarter:<10} ${y2024:>10,}K ${y2025:>10,}K ${growth_d:>10,}K {growth_p:>9.1f}%")

# Overall
print("-"*70)
total_2024 = sales_2024.sum()
total_2025 = sales_2025.sum()
total_growth = total_2025 - total_2024
total_growth_pct = (total_growth / total_2024) * 100
print(f"{'TOTAL':<10} ${total_2024:>10,}K ${total_2025:>10,}K ${total_growth:>10,}K {total_growth_pct:>9.1f}%")

# Best performers
print("\n--- Top Performers ---")
region_growth = sales_2025.sum(axis=1) - sales_2024.sum(axis=1)
best_region = regions[region_growth.argmax()]
print(f"Fastest Growing Region: {best_region} (+${region_growth.max()}K)")

quarter_growth = sales_2025.sum(axis=0) - sales_2024.sum(axis=0)
best_quarter = quarters[quarter_growth.argmax()]
print(f"Strongest Quarter: {best_quarter} (+${quarter_growth.max()}K)")
```
