```python
# Import NumPy
import numpy as np

# Set random seed for reproducibility (so everyone gets the same data)
np.random.seed(4641)

# Generate simulated daily returns data
# 20 stocks, 252 trading days (1 year)
# Returns are in decimal form (e.g., 0.02 = 2% return)

stock_names = np.array([
    'APEX', 'BOLT', 'CYBER', 'DATA', 'EDGE',
    'FLUX', 'GRID', 'HYPER', 'INTEL', 'JADE',
    'KITE', 'LOGIC', 'MATRIX', 'NEXUS', 'ORBIT',
    'PULSE', 'QUANT', 'RELAY', 'SPARK', 'TITAN'
])

# Generate returns with different volatility profiles
# High-growth tech stocks have higher volatility
volatilities = np.array([0.02, 0.025, 0.03, 0.018, 0.022,
                         0.028, 0.015, 0.035, 0.02, 0.025,
                         0.032, 0.019, 0.024, 0.021, 0.027,
                         0.023, 0.031, 0.017, 0.026, 0.029])

# Expected annual returns (varying by stock)
expected_returns = np.array([0.12, 0.15, 0.18, 0.10, 0.14,
                             0.16, 0.08, 0.22, 0.11, 0.13,
                             0.20, 0.09, 0.15, 0.12, 0.17,
                             0.14, 0.19, 0.07, 0.16, 0.21])

# Generate 252 days of returns for each stock
# Shape: (252, 20) - rows are days, columns are stocks
daily_returns = np.random.normal(
    loc=expected_returns/252,  # Daily expected return
    scale=volatilities,        # Daily volatility
    size=(252, 20)
)

# Portfolio weights (how much is invested in each stock)
# These sum to 1.0 (100% of portfolio)
weights = np.array([0.08, 0.06, 0.07, 0.05, 0.04,
                    0.05, 0.04, 0.06, 0.05, 0.04,
                    0.05, 0.04, 0.06, 0.05, 0.05,
                    0.05, 0.06, 0.04, 0.03, 0.03])

print("Data loaded successfully!")
print(f"Daily returns shape: {daily_returns.shape}")
print(f"Number of stocks: {len(stock_names)}")
print(f"Trading days: {daily_returns.shape[0]}")
print(f"Portfolio weights sum: {weights.sum():.2f}")
```

## Part 1: Individual Stock Analysis (20 points)

Sarah needs to understand the risk profile of each stock in the portfolio.

**Your Task:**
Using NumPy array operations (not loops), calculate for each stock:
1. Average daily return
2. Daily return standard deviation (volatility)
3. Annualized return (average daily return × 252)
4. Annualized volatility (daily std × √252)
5. Sharpe Ratio: (Annualized Return - Risk-Free Rate) / Annualized Volatility
   - Use risk-free rate = 5% (0.05)

**Identify:**
- The stock with the highest annualized return
- The stock with the lowest volatility
- The stock with the best risk-adjusted return (highest Sharpe ratio)

```python
# Part 1: Individual Stock Analysis
# Use NumPy operations - NO LOOPS

risk_free_rate = 0.05

# Calculate statistics for each stock
# Hint: Use axis parameter to operate across the correct dimension

# Average daily returns (one value per stock)
avg_daily_returns = np.mean(daily_returns, axis=0);

# Daily standard deviation
daily_std_dev = np.std(daily_returns, axis=0);

# Annualized return
annualized_returns = avg_daily_returns * 252;

# Annualized volatility
annualized_volatility = daily_std_dev * np.sqrt(252);

# Sharpe Ratio
sharpe_ratios = (annualized_returns - risk_free_rate) / annualized_volatility;

# Display results
print("INDIVIDUAL STOCK ANALYSIS")
print("="*70)
print(f"{'Stock':<8} {'Ann. Return':>12} {'Ann. Vol':>12} {'Sharpe':>10}")
print("-"*70)

# Print each stock's metrics using NumPy operations
def format_stock_row(stock, ann_ret, ann_vol, sharpe):
    return f"{stock:<8} {ann_ret:>12.4f} {ann_vol:>12.4f} {sharpe:>10.4f}"

vectorized_format_row = np.vectorize(format_stock_row)
formatted_stock_rows = vectorized_format_row(stock_names, annualized_returns, annualized_volatility, sharpe_ratios)
print('\n'.join(formatted_stock_rows))


# Identify best performers
print("\n" + "="*70)
print("KEY FINDINGS")
print("-"*70)

# Identify stock with the highest annualized return
highest_return_idx = np.argmax(annualized_returns)
highest_return_stock = stock_names[highest_return_idx]
print(f"Highest Annualized Return: {highest_return_stock} ({annualized_returns[highest_return_idx]:.4f})")

# Identify stock with the lowest volatility
lowest_volatility_idx = np.argmin(annualized_volatility)
lowest_volatility_stock = stock_names[lowest_volatility_idx]
print(f"Lowest Volatility: {lowest_volatility_stock} ({annualized_volatility[lowest_volatility_idx]:.4f})")

# Identify stock with the best risk-adjusted return (highest Sharpe ratio)
highest_sharpe_idx = np.argmax(sharpe_ratios)
highest_sharpe_stock = stock_names[highest_sharpe_idx]
print(f"Best Sharpe Ratio: {highest_sharpe_stock} ({sharpe_ratios[highest_sharpe_idx]:.4f})")
```

## Part 2: Portfolio Return Analysis (20 points)

Now Jordan needs to analyze the portfolio as a whole, considering how the weights affect overall performance.

**Your Task:**
1. Calculate the daily portfolio returns (weighted sum of individual returns)
   - Hint: For each day, multiply each stock's return by its weight and sum
2. Calculate the portfolio's:
   - Average daily return
   - Daily volatility
   - Annualized return
   - Annualized volatility
   - Sharpe ratio
3. Find the best and worst trading days
4. Calculate cumulative returns (how $1 invested would grow)

```python
# Part 2: Portfolio Return Analysis

# Calculate daily portfolio returns
# Hint: Use matrix multiplication or broadcasting with weights
portfolio_daily_returns = daily_returns @ weights

# Calculate portfolio statistics
portfolio_avg_daily_return = np.mean(portfolio_daily_returns)
portfolio_daily_volatility = np.std(portfolio_daily_returns)
portfolio_annualized_return = portfolio_avg_daily_return * 252
portfolio_annualized_volatility = portfolio_daily_volatility * np.sqrt(252)
portfolio_sharpe_ratio = (portfolio_annualized_return - risk_free_rate) / portfolio_annualized_volatility

# Find best and worst days
best_day_return_value = np.max(portfolio_daily_returns)
worst_day_return_value = np.min(portfolio_daily_returns)
best_day_index = np.argmax(portfolio_daily_returns) + 1 # +1 to make it 1-indexed for human readability
worst_day_index = np.argmin(portfolio_daily_returns) + 1 # +1 to make it 1-indexed for human readability

# Calculate cumulative returns
# If you invested $1 on day 0, what would it be worth each day?
# Cumulative return = cumprod(1 + daily_returns)
cumulative_returns = np.cumprod(1 + portfolio_daily_returns)
final_portfolio_value = cumulative_returns[-1]

# Display portfolio summary
print("PORTFOLIO PERFORMANCE SUMMARY")
print("="*60)
print(f"Annualized Return: {portfolio_annualized_return:.4f}")
print(f"Annualized Volatility: {portfolio_annualized_volatility:.4f}")
print(f"Sharpe Ratio: {portfolio_sharpe_ratio:.4f}")
print(f"\nBest Day (Day {best_day_index}): {best_day_return_value:.4f}")
print(f"Worst Day (Day {worst_day_index}): {worst_day_return_value:.4f}")
print(f"\nStarting Value: $1.00")
print(f"Ending Value: ${final_portfolio_value:.2f}")
```

## Part 3: Risk Metrics (25 points)

The investment committee is particularly interested in downside risk—how bad could things get?

**Your Task:**
Calculate the following risk metrics:

1. **Value at Risk (VaR) at 95% confidence**
   - What is the 5th percentile of daily returns?
   - Interpretation: "On 95% of days, the portfolio won't lose more than X%"

2. **Conditional VaR (Expected Shortfall)**
   - Average of returns that are worse than the VaR
   - Interpretation: "When losses exceed VaR, the average loss is X%"

3. **Maximum Drawdown**
   - The largest peak-to-trough decline in cumulative value
   - Hint: Calculate running maximum of cumulative returns, then find biggest drop

4. **Downside Deviation**
   - Standard deviation of only the negative returns
   - More focused measure of "bad" volatility

5. **Sortino Ratio**
   - Like Sharpe, but uses downside deviation instead of total volatility
   - (Annualized Return - Risk-Free Rate) / Annualized Downside Deviation

```python
# Part 3: Risk Metrics

# 1. Value at Risk (VaR) at 95% confidence
# Hint: Use np.percentile()
var_95 = np.percentile(portfolio_daily_returns, 5)
# Interpretation: "On 95% of days, the portfolio won't lose more than X%"

# 2. Conditional VaR (Expected Shortfall)
# Average of returns that are worse than the VaR
es_returns = portfolio_daily_returns[portfolio_daily_returns < var_95]
es_95 = np.mean(es_returns)
# Interpretation: "When losses exceed VaR, the average loss is X%"

# 3. Maximum Drawdown
# The largest peak-to-trough decline in cumulative value
# Hint: Calculate running maximum of cumulative returns, then find biggest drop
cum_returns_with_initial = np.insert(cumulative_returns, 0, 1.0) # Start with 1.0 for initial investment
peak = np.maximum.accumulate(cum_returns_with_initial)
drawdown = (peak - cum_returns_with_initial) / peak
max_drawdown = np.max(drawdown)

# 4. Downside Deviation
# Standard deviation of only the negative returns
# More focused measure of "bad" volatility
negative_returns = portfolio_daily_returns[portfolio_daily_returns < 0]
downside_deviation = np.std(negative_returns) if len(negative_returns) > 0 else 0
annualized_downside_deviation = downside_deviation * np.sqrt(252)

# 5. Sortino Ratio
# Like Sharpe, but uses downside deviation instead of total volatility
# (Annualized Return - Risk-Free Rate) / Annualized Downside Deviation
sortino_ratio = (portfolio_annualized_return - risk_free_rate) / annualized_downside_deviation if annualized_downside_deviation > 0 else np.nan

# Display risk report
print("RISK ANALYSIS REPORT")
print("="*60)
print("\nDownside Risk Metrics:")
print("-"*60)
print(f"Value at Risk (95%): {var_95:.4f}")
print(f"  → On 95% of days, the portfolio won't lose more than {abs(var_95):.2%} (or {var_95:.4f})")
print(f"\nConditional VaR (Expected Shortfall): {es_95:.4f}")
print(f"  → When losses exceed VaR, the average loss is {abs(es_95):.2%} (or {es_95:.4f})")
print(f"\nMaximum Drawdown: {max_drawdown:.4f}")
print(f"  → Largest peak-to-trough decline over the period")
print(f"\nDownside Deviation (annualized): {annualized_downside_deviation:.4f}")
print(f"\nSortino Ratio: {sortino_ratio:.4f}")
print(f"  → Risk-adjusted return using only downside volatility")
```

## Part 4: Weight Optimization Scenario (20 points)

Sarah asks Jordan to explore what would happen with different portfolio weights.

**Scenario: Equal Weight Portfolio**
What if instead of the current weights, we invested equally in all 20 stocks (5% each)?

**Your Task:**
1. Create equal weights (5% each stock)
2. Calculate the equal-weight portfolio's daily returns
3. Compare the equal-weight vs current portfolio:
   - Annualized return
   - Annualized volatility
   - Sharpe ratio
   - Maximum drawdown
4. Identify which stocks are overweight and underweight in the current portfolio compared to equal weight
5. Provide a brief interpretation: Which portfolio is better and why?

```python
# Part 4: Weight Optimization Scenario

# Create equal weights
num_stocks = len(stock_names)
equal_weights = np.full(num_stocks, 1/num_stocks) # 1/20 = 0.05

print("Equal Weights:")
print(equal_weights)
print(f"Sum of Equal Weights: {equal_weights.sum():.2f}")

# Calculate equal-weight portfolio returns
equal_weight_portfolio_daily_returns = daily_returns @ equal_weights

# Calculate equal-weight portfolio statistics
equal_weight_portfolio_avg_daily_return = np.mean(equal_weight_portfolio_daily_returns)
equal_weight_portfolio_daily_volatility = np.std(equal_weight_portfolio_daily_returns)
equal_weight_portfolio_annualized_return = equal_weight_portfolio_avg_daily_return * 252
equal_weight_portfolio_annualized_volatility = equal_weight_portfolio_daily_volatility * np.sqrt(252)
equal_weight_portfolio_sharpe_ratio = (equal_weight_portfolio_annualized_return - risk_free_rate) / equal_weight_portfolio_annualized_volatility

# Calculate cumulative returns for equal-weight portfolio
equal_weight_cumulative_returns = np.cumprod(1 + equal_weight_portfolio_daily_returns)

# Calculate maximum drawdown for equal-weight portfolio
equal_weight_cum_returns_with_initial = np.insert(equal_weight_cumulative_returns, 0, 1.0)
equal_weight_peak = np.maximum.accumulate(equal_weight_cum_returns_with_initial)
equal_weight_drawdown = (equal_weight_peak - equal_weight_cum_returns_with_initial) / equal_weight_peak
equal_weight_max_drawdown = np.max(equal_weight_drawdown)

# Store current portfolio metrics in variables
current_annualized_return = portfolio_annualized_return
current_annualized_volatility = portfolio_annualized_volatility
current_sharpe_ratio = portfolio_sharpe_ratio
current_max_drawdown = max_drawdown

# Compare portfolios
print("PORTFOLIO COMPARISON")
print("="*60)
print(f"{'Metric':<25} {'Current':>15} {'Equal Weight':>15}")
print("-"*60)
# Print comparison table
print(f"{'Annualized Return':<25} {current_annualized_return:>15.4f} {equal_weight_portfolio_annualized_return:>15.4f}")
print(f"{'Annualized Volatility':<25} {current_annualized_volatility:>15.4f} {equal_weight_portfolio_annualized_volatility:>15.4f}")
print(f"{'Sharpe Ratio':<25} {current_sharpe_ratio:>15.4f} {equal_weight_portfolio_sharpe_ratio:>15.4f}")
print(f"{'Maximum Drawdown':<25} {current_max_drawdown:>15.4f} {equal_weight_max_drawdown:>15.4f}")

# Identify over/underweight positions
print("\nWEIGHT DIFFERENCES (Current vs Equal)")
print("-"*60)

# Calculate the difference between the current portfolio's weights and the equal_weights
weight_differences = weights - equal_weights

# Identify stocks where the current weight is greater than the equal weight (overweight)
overweight_indices = np.where(weight_differences > 0)
overweight_stocks = stock_names[overweight_indices]
overweight_diffs = weight_differences[overweight_indices]

# Identify stocks where the current weight is less than the equal weight (underweight)
underweight_indices = np.where(weight_differences < 0)
underweight_stocks = stock_names[underweight_indices]
underweight_diffs = weight_differences[underweight_indices]

print("Overweight Stocks:")
if len(overweight_stocks) > 0:
    for stock, diff in zip(overweight_stocks, overweight_diffs):
        print(f"  {stock:<10}: {diff:>8.2%}")
else:
    print("  None")

print("\nUnderweight Stocks:")
if len(underweight_stocks) > 0:
    for stock, diff in zip(underweight_stocks, underweight_diffs):
        print(f"  {stock:<10}: {diff:>8.2%}")
else:
    print("  None")

# Interpretation
print("\nINTERPRETATION")
print("-"*60)
print("Based on the comparison:")
print("The Equal Weight Portfolio shows a higher Annualized Return (0.2151 vs 0.1981) and a higher Sharpe Ratio (1.8180 vs 1.5347),")
print("indicating better risk-adjusted returns compared to the Current Portfolio. It also has lower Annualized Volatility (0.0908 vs 0.0965).")
print("However, the Equal Weight Portfolio has a slightly higher Maximum Drawdown (0.0659 vs 0.0591),")
print("suggesting it experienced a slightly larger peak-to-trough decline during the period.")
print("Overall, the Equal Weight Portfolio appears to be a more efficient portfolio in terms of return per unit of risk (Sharpe Ratio).")
```
