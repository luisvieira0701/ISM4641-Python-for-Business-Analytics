## Part 1: Shipping Cost Calculator (25 points)

The most inconsistent calculation at SwiftShip is the basic shipping cost. Priya documents the official pricing rules:

**Base Rate by Zone:**

  Zone 1 (Local, < 100 miles): $5.99

  Zone 2 (Regional, 100-500 miles): $9.99

  Zone 3 (National, 500-1500 miles): $14.99

  Zone 4 (Extended, > 1500 miles): $24.99

**Weight Surcharge:**
- First 5 lbs: Included in base rate

- 5-20 lbs: $0.50 per lb over 5

- 20-50 lbs: $0.75 per lb over 20 (plus the 5-20 charge)

- Over 50 lbs: $1.00 per lb over 50 (plus previous charges)

**Service Level Multiplier:**
- Standard (5-7 days): 1.0x
- Express (2-3 days): 1.5x
- Overnight: 2.5x

**Your Task:**
Create three functions:
1. `get_base_rate(distance_miles)` - Returns the base rate for a given distance
2. `calculate_weight_surcharge(weight_lbs)` - Returns the weight surcharge
3. `calculate_shipping_cost(distance_miles, weight_lbs, service_level="standard")` - Returns the total shipping cost

The third function should call the first two.

```python
import math

# Part 1: Shipping Cost Calculator
# Build the core pricing functions

def get_base_rate(distance_miles):
    """
    Determine the base shipping rate based on distance.

    Parameters:
        distance_miles (float): The shipping distance in miles

    Returns:
        float: The base shipping rate in dollars
    """
    if distance_miles < 100:
        return 5.99  # Zone 1 (Local, < 100 miles)
    elif 100 <= distance_miles < 500:
        return 9.99  # Zone 2 (Regional, 100-500 miles)
    elif 500 <= distance_miles < 1500:
        return 14.99 # Zone 3 (National, 500-1500 miles)
    else: # distance_miles >= 1500
        return 24.99 # Zone 4 (Extended, > 1500 miles)


def calculate_weight_surcharge(weight_lbs):
    """
    Calculate the weight surcharge based on package weight.

    Parameters:
        weight_lbs (float): The package weight in pounds

    Returns:
        float: The weight surcharge in dollars
    """
    surcharge = 0.0
    if weight_lbs <= 5:
        surcharge = 0.0
    elif 5 < weight_lbs <= 20:
        surcharge = (weight_lbs - 5) * 0.50
    elif 20 < weight_lbs <= 50:
        # Surcharge for 5-20 lbs (15 lbs * $0.50) + surcharge for weight over 20 lbs
        surcharge = (15 * 0.50) + ((weight_lbs - 20) * 0.75)
    else: # weight_lbs > 50
        # Surcharge for 5-20 lbs (15 lbs * $0.50)
        # + surcharge for 20-50 lbs (30 lbs * $0.75)
        # + surcharge for weight over 50 lbs
        surcharge = (15 * 0.50) + (30 * 0.75) + ((weight_lbs - 50) * 1.00)
    return surcharge


def calculate_shipping_cost(distance_miles, weight_lbs, service_level="standard"):
    """
    Calculate the total shipping cost for a package.

    Parameters:
        distance_miles (float): The shipping distance in miles
        weight_lbs (float): The package weight in pounds
        service_level (str): The service level - 'standard', 'express', or 'overnight'

    Returns:
        float: The total shipping cost in dollars
    """
    base_rate = get_base_rate(distance_miles)
    weight_surcharge = calculate_weight_surcharge(weight_lbs)

    total_cost = base_rate + weight_surcharge

    # Apply service level multiplier
    multiplier = 1.0 # standard service level
    if service_level.lower() == "express":
        multiplier = 1.5
    elif service_level.lower() == "overnight":
        multiplier = 2.5
    # For 'standard', multiplier remains 1.0 (default)

    return total_cost * multiplier
```

## Part 2: Delivery Time Estimator (20 points)

Customers always want to know: "When will my package arrive?" SwiftShip needs a consistent way to estimate delivery times.

**Business Days Calculation:**
- Standard: 5-7 business days → use 6 as estimate
- Express: 2-3 business days → use 2 as estimate  
- Overnight: 1 business day

**Additional Delays:**
- Distances over 1000 miles: Add 1 business day
- Packages over 50 lbs: Add 1 business day (requires special handling)
- Rural delivery (is_rural=True): Add 2 business days

**Your Task:**
Create a function `estimate_delivery(distance_miles, weight_lbs, service_level, is_rural=False)` that:
1. Calculates the estimated business days for delivery
2. Returns both the minimum and maximum days as a tuple: `(min_days, max_days)`
   - For standard: range is estimate ± 1
   - For express: range is estimate to estimate + 1
   - For overnight: range is estimate to estimate

```python
# Part 2: Delivery Time Estimator

def estimate_delivery(distance_miles, weight_lbs, service_level, is_rural=False):
    """
    Estimate the delivery time window for a shipment.

    Parameters:
        distance_miles (float): The shipping distance in miles
        weight_lbs (float): The package weight in pounds
        service_level (str): The service level - 'standard', 'express', or 'overnight'
        is_rural (bool): Whether the delivery address is in a rural area

    Returns:
        tuple: (min_days, max_days) representing the delivery window
    """
    initial_base_estimate = 0

    # Step 1: Initial Business Days Calculation based on Service Level
    # Note: These estimates reflect typical transit times agreed upon with carriers.
    if service_level.lower() == "standard":
        initial_base_estimate = 6 # Standard: 5-7 business days, using 6 as the estimate
    elif service_level.lower() == "express":
        initial_base_estimate = 2 # Express: 2-3 business days, using 2 as the estimate
    elif service_level.lower() == "overnight":
        initial_base_estimate = 1 # Overnight: 1 business day
    else:
        # Default to standard if an unrecognized service level is provided
        initial_base_estimate = 6

    # Step 2: Calculate Additional Delays
    total_delays = 0
    if distance_miles > 1000:
        total_delays += 1
    if weight_lbs > 50:
        total_delays += 1
        # Note for stakeholders: Packages over 50 lbs require special handling, ensuring careful transport.
    if is_rural:
        total_delays += 2

    # Step 3: Determine min and max days for the return tuple based on service level rules
    min_days_final = 0
    max_days_final = 0

    if service_level.lower() == "standard":
        # For standard, 'estimate' includes initial base + all delays
        effective_estimate = initial_base_estimate + total_delays
        min_days_final = max(1, effective_estimate - 1) # Ensure min_days is at least 1
        max_days_final = effective_estimate + 1
    elif service_level.lower() == "express":
        # For express, min_days is the initial base estimate, delays extend max_days
        min_days_final = initial_base_estimate
        max_days_final = (initial_base_estimate + total_delays) + 1
    elif service_level.lower() == "overnight":
        # For overnight, 'estimate' includes initial base + all delays
        effective_estimate = initial_base_estimate + total_delays
        min_days_final = effective_estimate
        max_days_final = effective_estimate
    else:
        # Default to standard if unrecognized service level
        effective_estimate = initial_base_estimate + total_delays
        min_days_final = max(1, effective_estimate - 1)
        max_days_final = effective_estimate + 1

    return (min_days_final, max_days_final)
```

## Part 3: Customer Quote Generator (25 points)

Now Priya needs to combine everything into a professional quote generator that the sales team can use.

**Quote Requirements:**
- Calculate shipping cost
- Calculate estimated delivery window
- Apply volume discount if quantity > 10 packages:
  - 11-25 packages: 5% discount
  - 26-50 packages: 10% discount
  - 51+ packages: 15% discount
- Return a dictionary containing all quote details

**Your Task:**
Create a function `generate_quote(distance, weight, service_level, quantity, is_rural=False)` that:
1. Uses your previous functions to calculate costs and delivery times
2. Applies volume discounts
3. Returns a dictionary with all quote information

```python
# Part 3: Customer Quote Generator

def calculate_volume_discount(quantity):
    """
    Calculate the volume discount rate based on package quantity.

    Parameters:
        quantity (int): Number of packages

    Returns:
        float: Discount rate as a decimal (e.g., 0.10 for 10%)
    """
    if 11 <= quantity <= 25:
        return 0.05  # 5% discount
    elif 26 <= quantity <= 50:
        return 0.10  # 10% discount
    elif quantity >= 51:
        return 0.15  # 15% discount
    else:
        return 0.0   # No discount for 10 or fewer packages


def generate_quote(distance, weight, service_level, quantity, is_rural=False):
    """
    Generate a complete shipping quote for a customer.

    Parameters:
        distance (float): Shipping distance in miles
        weight (float): Package weight in pounds
        service_level (str): Service level ('standard', 'express', 'overnight')
        quantity (int): Number of packages
        is_rural (bool): Whether delivery is to a rural area

    Returns:
        dict: Quote details including costs, discounts, and delivery estimate
    """
    # Calculate unit cost per package
    unit_cost = calculate_shipping_cost(distance, weight, service_level)

    # Calculate volume discount
    discount_rate = calculate_volume_discount(quantity)

    # Calculate subtotal before discount
    subtotal = unit_cost * quantity

    # Calculate discount amount
    discount_amount = subtotal * discount_rate

    # Calculate final total after discount
    total = subtotal - discount_amount

    # Estimate delivery time
    delivery_min, delivery_max = estimate_delivery(distance, weight, service_level, is_rural)

    # Assemble quote details into a dictionary
    quote_details = {
        "unit_cost": unit_cost,
        "quantity": quantity,
        "subtotal": subtotal,
        "discount_rate": discount_rate,
        "discount_amount": discount_amount,
        "total": total,
        "delivery_min": delivery_min,
        "delivery_max": delivery_max
    }

    return quote_details
```

## Part 4: Batch Processing (15 points)

SwiftShip receives bulk orders from business customers. Priya needs a function to process multiple shipments at once and generate a summary report.

**Your Task:**
Create a function `process_batch(shipments)` that:
1. Takes a list of shipment dictionaries as input
2. Generates a quote for each shipment
3. Returns a summary dictionary containing:
   - Total number of packages across all shipments
   - Total cost before discounts
   - Total discounts applied
   - Grand total after discounts
   - List of all individual quotes

```python
# Part 4: Batch Processing

def process_batch(shipments):
    """
    Process a batch of shipments and generate a summary.

    Parameters:
        shipments (list): List of dictionaries, each containing:
            - distance (float)
            - weight (float)
            - service_level (str)
            - quantity (int)
            - is_rural (bool, optional)

    Returns:
        dict: Batch summary with totals and individual quotes
    """
    total_packages = 0
    total_subtotal = 0.0
    total_discounts = 0.0
    grand_total = 0.0
    individual_quotes = []

    for shipment in shipments:
        # Extract parameters, providing default for is_rural if not present
        distance = shipment['distance']
        weight = shipment['weight']
        service_level = shipment['service_level']
        quantity = shipment['quantity']
        is_rural = shipment.get('is_rural', False) # Default to False if not specified

        # Generate quote for the current shipment
        quote = generate_quote(distance, weight, service_level, quantity, is_rural)
        individual_quotes.append(quote)

        # Aggregate summary data
        total_packages += quote['quantity']
        total_subtotal += quote['subtotal']
        total_discounts += quote['discount_amount']
        grand_total += quote['total']

    # Assemble the batch summary dictionary
    batch_summary = {
        "total_packages": total_packages,
        "total_subtotal": total_subtotal,
        "total_discounts": total_discounts,
        "grand_total": grand_total,
        "individual_quotes": individual_quotes
    }

    return batch_summary
```

## Part 5: Documentation and Testing (15 points)

Before presenting to Marcus, Priya needs to create documentation showing that the new system is reliable and consistent.

**Your Task:**
1. Create a function `run_validation_tests()` that tests edge cases:
   - Minimum distance (1 mile)
   - Maximum weight (100 lbs)
   - All three service levels
   - Rural vs non-rural
   - Single package vs bulk order

2. Print a formatted report showing that all tests pass with expected behavior.

3. Create a `print_function_documentation()` function that prints a summary of all the functions you created, their purposes, and their parameters.

```python
import math

# Part 5: Documentation and Testing

def run_validation_tests():
    """
    Run a comprehensive set of validation tests on the pricing system.
    Prints results and returns True if all tests pass.
    """
    print("SWIFTSHIP PRICING SYSTEM - VALIDATION REPORT")
    print("="*60)

    all_passed = True

    # Helper function for assertion and printing results (for costs $)
    def assert_pricing_system_test(test_name, expected, actual):
        nonlocal all_passed
        # Use a small tolerance for float comparisons
        if abs(expected - actual) < 0.01:
            print(f"[PASS] {test_name}: Expected ${expected:.2f}, Got ${actual:.2f}")
        else:
            print(f"[FAIL] {test_name}: Expected ${expected:.2f}, Got ${actual:.2f}")
            all_passed = False

    # Function to test delivery ranges
    def assert_delivery_range_test(test_name, expected_min, expected_max, actual_min, actual_max):
        nonlocal all_passed
        if expected_min == actual_min and expected_max == actual_max:
            print(f"[PASS] {test_name}: Expected {expected_min}-{expected_max} days, Got {actual_min}-{actual_max} days")
        else:
            print(f"[FAIL] {test_name}: Expected {expected_min}-{expected_max} days, Got {actual_min}-{actual_max} days")
            all_passed = False

    # Test 1: Minimum distance (1 mile), 3 lbs, standard
    cost_min_dist = calculate_shipping_cost(1, 3, "standard")
    assert_pricing_system_test("Min Distance (1 mi, 3 lbs, standard)", 5.99, cost_min_dist)

    # Test 2: Maximum weight (100 lbs), 100 miles, standard
    # Expected: base rate $9.99, weight surcharge $80.00, total $89.99
    cost_max_weight = calculate_shipping_cost(100, 100, "standard")
    assert_pricing_system_test("Max Weight (100 mi, 100 lbs, standard)", 89.99, cost_max_weight)

    # Test 3: All service levels (500 miles, 10 lbs)
    # Base rate for 500 miles is $14.99
    # Weight surcharge for 10 lbs is (10-5) * 0.50 = $2.50
    # Total cost before multiplier = 14.99 + 2.50 = 17.49
    base_cost_for_service_tests = 17.49

    cost_standard = calculate_shipping_cost(500, 10, "standard") # Expected 17.49 * 1.0 = 17.49
    assert_pricing_system_test("Service Level (500 mi, 10 lbs, standard)", base_cost_for_service_tests * 1.0, cost_standard)

    cost_express = calculate_shipping_cost(500, 10, "express")   # Expected 17.49 * 1.5 = 26.235
    assert_pricing_system_test("Service Level (500 mi, 10 lbs, express)", base_cost_for_service_tests * 1.5, cost_express)

    cost_overnight = calculate_shipping_cost(500, 10, "overnight") # Expected 17.49 * 2.5 = 43.725
    assert_pricing_system_test("Service Level (500 mi, 10 lbs, overnight)", base_cost_for_service_tests * 2.5, cost_overnight)

    # Test 4: Rural vs non-rural (Delivery Estimation)
    # 200 mi, 10 lbs, standard: Base estimate 6, no delays for non-rural -> 5-7 days
    min_nr, max_nr = estimate_delivery(200, 10, "standard", False)
    assert_delivery_range_test("Delivery (200 mi, 10 lbs, standard, Non-Rural)", 5, 7, min_nr, max_nr)

    # 200 mi, 10 lbs, standard, rural: Base estimate 6, +2 for rural -> effective estimate 8 -> 7-9 days
    min_r, max_r = estimate_delivery(200, 10, "standard", True)
    assert_delivery_range_test("Delivery (200 mi, 10 lbs, standard, Rural)", 7, 9, min_r, max_r)

    # Test 5: Volume discounts (Generate Quote)
    # Single package (no discount) -> total should be unit_cost
    unit_cost_single = calculate_shipping_cost(100, 10, "standard") # 9.99 base + 2.50 weight surcharge = 12.49
    quote_single = generate_quote(100, 10, "standard", 1, False)
    assert_pricing_system_test("Volume Discount (1 package)", unit_cost_single, quote_single['total'])

    # 20 packages (5% discount)
    unit_cost_20 = calculate_shipping_cost(100, 10, "standard") # 12.49
    expected_total_20 = (unit_cost_20 * 20) * (1 - 0.05)
    quote_20 = generate_quote(100, 10, "standard", 20, False)
    assert_pricing_system_test("Volume Discount (20 packages, 5%)", expected_total_20, quote_20['total'])

    # 60 packages (15% discount)
    unit_cost_60 = calculate_shipping_cost(100, 10, "standard") # 12.49
    expected_total_60 = (unit_cost_60 * 60) * (1 - 0.15)
    quote_60 = generate_quote(100, 10, "standard", 60, False)
    assert_pricing_system_test("Volume Discount (60 packages, 15%)", expected_total_60, quote_60['total'])

    print("\n" + "="*60)
    if all_passed:
        print("ALL TESTS PASSED - System ready for deployment")
    else:
        print("SOME TESTS FAILED - Review required")

    return all_passed


def print_function_documentation():
    """
    Print documentation for all functions in the pricing library.
    """
    print("\nFUNCTION DOCUMENTATION")
    print("="*60)

    print("\n### Part 1: Shipping Cost Calculator")
    print("These functions handle the core calculations for how much it costs to ship a package, based on distance, weight, and desired speed.")
    print("  - **`get_base_rate(distance_miles)`:** Determines the initial shipping fee based purely on how far the package needs to travel. (Parameters: `distance_miles` - how many miles)")
    print("  - **`calculate_weight_surcharge(weight_lbs)`:** Calculates additional charges for heavier packages, ensuring we cover the extra handling. (Parameters: `weight_lbs` - package weight in pounds)")
    print("  - **`calculate_shipping_cost(distance_miles, weight_lbs, service_level)`:** Combines the base rate and weight surcharge, then adjusts for how quickly the customer wants it delivered (Standard, Express, Overnight). (Parameters: `distance_miles`, `weight_lbs`, `service_level` - speed of delivery)")

    print("\n### Part 2: Delivery Time Estimator")
    print("This function provides customers with a realistic window for when their package will arrive, considering various factors.")
    print("  - **`estimate_delivery(distance_miles, weight_lbs, service_level, is_rural)`:** Gives an estimated delivery timeframe in business days, taking into account distance, weight, service speed, and if the delivery location is rural. (Parameters: `distance_miles`, `weight_lbs`, `service_level`, `is_rural` - whether it's a rural address)")

    print("\n### Part 3: Customer Quote Generator")
    print("These functions are used by the sales team to create comprehensive and accurate quotes for customers, including any applicable discounts.")
    print("  - **`calculate_volume_discount(quantity)`:** Figures out if a customer qualifies for a discount based on the number of packages they are shipping. (Parameters: `quantity` - number of packages)")
    print("  - **`generate_quote(distance, weight, service_level, quantity, is_rural)`:** Produces a full quote detailing costs, discounts, and delivery estimates for a shipment. (Parameters: `distance`, `weight`, `service_level`, `quantity`, `is_rural`)")

    print("\n### Part 4: Batch Processing")
    print("This function helps us manage large orders efficiently by processing multiple shipments at once and summarizing the total costs.")
    print("  - **`process_batch(shipments)`:** Takes a list of many shipments, generates a quote for each, and provides an overall summary of total packages, costs, and discounts for the entire batch. (Parameters: `shipments` - a list of individual shipment details)")

    print("\n### Part 5: Documentation and Testing")
    print("These functions ensure the reliability and clarity of our pricing system.")
    print("  - **`run_validation_tests()`:** Automatically checks our pricing system against various scenarios to ensure all calculations are correct and consistent. (No specific parameters needed)")
    print("  - **`print_function_documentation()`:** (You are currently viewing this documentation!) Provides a clear overview of all the functions in our system, their purpose, and what information they need. (No specific parameters needed)")
```
