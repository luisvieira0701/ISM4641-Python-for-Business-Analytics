## 4.1 Understanding Functions: The Recipe Analogy

Think of a function like a recipe in a cookbook:

| Recipe Component | Function Equivalent | Example |
|------------------|--------------------|---------|
| Recipe Name | Function name | `calculate_tax` |
| Ingredients List | Parameters | `(amount, rate)` |
| Cooking Instructions | Function body | The code inside |
| Finished Dish | Return value | The calculated result |

Just like you don't rewrite a recipe every time you want to make cookies, you shouldn't rewrite code every time you need to perform a calculation.

### Why Functions Matter in Business

**Without functions:**
```python
# Calculate tax for customer 1
tax1 = 100.00 * 0.07
total1 = 100.00 + tax1

# Calculate tax for customer 2
tax2 = 250.00 * 0.07
total2 = 250.00 + tax2

# Calculate tax for customer 3... and so on for 1000 customers
```

**With functions:**
```python
def calculate_total_with_tax(amount, tax_rate=0.07):
    return amount * (1 + tax_rate)

# Now use it for any number of customers
total1 = calculate_total_with_tax(100.00)
total2 = calculate_total_with_tax(250.00)
# Easy to process 1000 customers in a loop!
```

The benefits are enormous:
- **DRY Principle**: "Don't Repeat Yourself"—write once, use everywhere
- **Single Point of Change**: Update the formula once, and all calculations use the new version
- **Testing**: Test the function once, be confident it works everywhere
- **Readability**: `calculate_total_with_tax(amount)` is clearer than `amount * 1.07`

## 4.2 Defining Functions: Syntax and Structure

### Basic Function Syntax

```python
def function_name(parameter1, parameter2, ...):
    """Docstring: explains what the function does."""
    # Function body - the code that runs
    # when the function is called
    return result  # Optional: send a value back
```

Let's break down each component:

| Component | Description | Required? |
|-----------|-------------|----------|
| `def` | Keyword that starts a function definition | Yes |
| `function_name` | Identifier for calling the function | Yes |
| `parameters` | Input values the function accepts | No |
| `docstring` | Documentation describing the function | Recommended |
| `function body` | Code that executes when called | Yes |
| `return` | Statement that sends a value back | No |

````python
# The simplest possible function - no parameters, no return
def greet():
    """Display a welcome message."""
    print("Welcome to Python for Business Analytics!")

# Calling the function
greet()
````

````python
# Function with one parameter
def greet_employee(name):
    """Display a personalized welcome message."""
    print(f"Welcome aboard, {name}! We're excited to have you on the team.")

# Call with different arguments
greet_employee("Alice")
greet_employee("Bob")
greet_employee("Carol")
````

````python
# Function with multiple parameters
def greet_with_title(name, title, department):
    """Display a formal welcome message with job details."""
    print(f"Welcome, {title} {name}!")
    print(f"You'll be joining our {department} department.")
    print()

greet_with_title("Smith", "Dr.", "Data Science")
greet_with_title("Johnson", "Ms.", "Marketing")
````

### Naming Conventions for Functions

Python follows specific conventions for function names (PEP 8 style guide):

| Convention | Example | Notes |
|------------|---------|-------|
| Use lowercase | `calculate_tax` | Not `CalculateTax` |
| Use underscores | `get_customer_data` | Not `getcustomerdata` |
| Use verbs | `calculate`, `get`, `validate` | Describes action |
| Be descriptive | `calculate_monthly_revenue` | Not `calc` or `cmr` |

**Good function names:**
- `calculate_total_price()`
- `validate_email_address()`
- `get_customer_orders()`
- `format_currency()`

**Poor function names:**
- `do_stuff()` - Too vague
- `CalculateTax()` - Wrong style (CamelCase is for classes)
- `x()` - Meaningless
- `process()` - What does it process?

## 4.3 Return Values: Getting Results from Functions

A function without a `return` statement is like a worker who does their job but never reports back. Sometimes that's fine (like printing a message), but usually we need the result of a calculation.

### Functions Without Return

Functions that don't explicitly return a value return `None`:

````python
def print_receipt(items, total):
    """Print a formatted receipt (no return value needed)."""
    print("="*30)
    print("       RECEIPT")
    print("="*30)
    for item, price in items:
        print(f"{item:20} ${price:7.2f}")
    print("-"*30)
    print(f"{'TOTAL':20} ${total:7.2f}")
    print("="*30)

# This function's purpose is to print, not return
purchases = [("Coffee", 4.50), ("Sandwich", 8.99), ("Cookie", 2.50)]
print_receipt(purchases, 15.99)

# What happens if we try to capture the return?
result = print_receipt(purchases, 15.99)
print(f"Return value: {result}")  # None!
````

### Functions That Return Values

Most business functions need to return calculated values:

````python
def calculate_sales_tax(amount, tax_rate):
    """Calculate sales tax on a purchase amount.
    
    Args:
        amount: The purchase amount before tax
        tax_rate: The tax rate as a decimal (e.g., 0.07 for 7%)
    
    Returns:
        The calculated tax amount
    """
    tax = amount * tax_rate
    return tax

# Using the returned value
purchase = 250.00
florida_tax_rate = 0.07

tax_amount = calculate_sales_tax(purchase, florida_tax_rate)
total = purchase + tax_amount

print(f"Purchase: ${purchase:.2f}")
print(f"Tax (7%): ${tax_amount:.2f}")
print(f"Total: ${total:.2f}")
````

````python
# Return values can be used directly in expressions
def calculate_discount(price, discount_percent):
    """Calculate the discount amount."""
    return price * (discount_percent / 100)

original_price = 100.00

# Use the return value directly in a calculation
final_price = original_price - calculate_discount(original_price, 20)
print(f"Original: ${original_price:.2f}")
print(f"After 20% off: ${final_price:.2f}")

# Or use it in a condition
if calculate_discount(original_price, 20) > 15:
    print("Great deal! You're saving more than $15!")
````

### Returning Multiple Values

Python functions can return multiple values as a tuple. This is incredibly useful for business calculations that produce several related results:

````python
def calculate_order_breakdown(subtotal, tax_rate=0.07, discount_percent=0):
    """Calculate a complete order breakdown.
    
    Returns:
        tuple: (discount_amount, tax_amount, total)
    """
    discount = subtotal * (discount_percent / 100)
    taxable_amount = subtotal - discount
    tax = taxable_amount * tax_rate
    total = taxable_amount + tax
    
    return discount, tax, total  # Returns a tuple

# Method 1: Unpack into separate variables
discount, tax, total = calculate_order_breakdown(200, discount_percent=15)

print("Order Summary")
print(f"Subtotal: $200.00")
print(f"Discount (15%): -${discount:.2f}")
print(f"Tax (7%): ${tax:.2f}")
print(f"Total: ${total:.2f}")
````

````python
# Method 2: Keep as tuple
result = calculate_order_breakdown(500, discount_percent=10)
print(f"Result tuple: {result}")
print(f"Discount: ${result[0]:.2f}")
print(f"Tax: ${result[1]:.2f}")
print(f"Total: ${result[2]:.2f}")
````

````python
# Returning a dictionary for named results (even clearer)
def calculate_loan_payment(principal, annual_rate, years):
    """Calculate monthly loan payment and total cost.
    
    Returns:
        dict: Contains monthly_payment, total_paid, total_interest
    """
    monthly_rate = annual_rate / 12
    num_payments = years * 12
    
    # Monthly payment formula
    monthly_payment = principal * (monthly_rate * (1 + monthly_rate)**num_payments) / \
                      ((1 + monthly_rate)**num_payments - 1)
    
    total_paid = monthly_payment * num_payments
    total_interest = total_paid - principal
    
    return {
        'monthly_payment': monthly_payment,
        'total_paid': total_paid,
        'total_interest': total_interest,
        'num_payments': num_payments
    }

# Calculate a car loan
loan_info = calculate_loan_payment(25000, 0.059, 5)

print("Car Loan Analysis")
print("="*35)
print(f"Principal: $25,000")
print(f"Interest Rate: 5.9%")
print(f"Term: 5 years ({loan_info['num_payments']} payments)")
print("-"*35)
print(f"Monthly Payment: ${loan_info['monthly_payment']:,.2f}")
print(f"Total Paid: ${loan_info['total_paid']:,.2f}")
print(f"Total Interest: ${loan_info['total_interest']:,.2f}")
````

### Early Return: Exiting Functions Early

The `return` statement immediately exits the function. This is useful for validation and handling special cases:

````python
def calculate_discount_price(price, discount_code):
    """Apply a discount code to a price.
    
    Discount codes:
        SAVE10 - 10% off
        SAVE20 - 20% off
        HALFOFF - 50% off
    """
    # Validate price
    if price <= 0:
        return 0  # Early return for invalid price
    
    # Convert code to uppercase for comparison
    code = discount_code.upper()
    
    # Check discount codes
    if code == "SAVE10":
        return price * 0.90
    elif code == "SAVE20":
        return price * 0.80
    elif code == "HALFOFF":
        return price * 0.50
    else:
        return price  # Invalid code, no discount

# Test different scenarios
print(f"$100 with SAVE10: ${calculate_discount_price(100, 'SAVE10'):.2f}")
print(f"$100 with SAVE20: ${calculate_discount_price(100, 'save20'):.2f}")  # lowercase works
print(f"$100 with HALFOFF: ${calculate_discount_price(100, 'HALFOFF'):.2f}")
print(f"$100 with INVALID: ${calculate_discount_price(100, 'INVALID'):.2f}")
print(f"$-50 with SAVE10: ${calculate_discount_price(-50, 'SAVE10'):.2f}")  # Invalid price
````

## 4.4 Function Parameters: Flexibility in Input

Python offers several ways to pass arguments to functions, each with its own use case.

### Positional Arguments

Arguments are matched to parameters by their position (order matters!):

````python
def calculate_profit_margin(revenue, cost):
    """Calculate profit margin percentage."""
    profit = revenue - cost
    margin = (profit / revenue) * 100
    return margin

# Correct usage: revenue first, then cost
margin = calculate_profit_margin(100000, 70000)
print(f"Profit margin: {margin:.1f}%")

# WRONG: Order matters!
wrong_margin = calculate_profit_margin(70000, 100000)  # Oops! Cost > Revenue
print(f"Wrong margin: {wrong_margin:.1f}%")  # Negative margin!
````

### Keyword Arguments

Explicitly name which parameter receives which value (order doesn't matter):

````python
# Using keyword arguments - much clearer!
margin1 = calculate_profit_margin(revenue=100000, cost=70000)
margin2 = calculate_profit_margin(cost=70000, revenue=100000)  # Order doesn't matter

print(f"Margin 1: {margin1:.1f}%")
print(f"Margin 2: {margin2:.1f}%")  # Same result!
````

````python
# Mixing positional and keyword (positional must come first)
def create_employee_record(name, department, salary, start_date):
    """Create an employee record dictionary."""
    return {
        'name': name,
        'department': department,
        'salary': salary,
        'start_date': start_date
    }

# Valid combinations
emp1 = create_employee_record("Alice", "Sales", 65000, "2026-01-15")
emp2 = create_employee_record("Bob", "IT", salary=75000, start_date="2026-02-01")
emp3 = create_employee_record("Carol", department="Marketing", salary=70000, start_date="2026-03-01")

for emp in [emp1, emp2, emp3]:
    print(f"{emp['name']}: {emp['department']}, ${emp['salary']:,}")
````

### Default Parameter Values

Provide fallback values for parameters that aren't always needed:

````python
def calculate_total_with_tax(amount, tax_rate=0.07, discount=0):
    """
    Calculate total price with tax and optional discount.
    
    Args:
        amount: Base price (required)
        tax_rate: Tax rate as decimal (default: 0.07 = 7%)
        discount: Discount amount in dollars (default: 0)
    """
    taxable = amount - discount
    tax = taxable * tax_rate
    return taxable + tax

# Different ways to call this function
print("Scenario 1: Just amount (use all defaults)")
print(f"  $100 purchase: ${calculate_total_with_tax(100):.2f}")

print("\nScenario 2: Custom tax rate")
print(f"  $100 purchase in California (7.25%): ${calculate_total_with_tax(100, 0.0725):.2f}")

print("\nScenario 3: Custom tax rate and discount")
print(f"  $100 purchase, 8% tax, $10 off: ${calculate_total_with_tax(100, 0.08, 10):.2f}")

print("\nScenario 4: Use keyword to skip tax_rate, just set discount")
print(f"  $100 purchase with $15 discount: ${calculate_total_with_tax(100, discount=15):.2f}")
````

### Important: Mutable Default Arguments Gotcha

Never use a mutable object (list, dictionary) as a default argument! Here's why:

````python
# WRONG WAY - Don't do this!
def add_item_wrong(item, shopping_list=[]):
    """BAD: Default list is shared between all calls!"""
    shopping_list.append(item)
    return shopping_list

# Watch what happens
list1 = add_item_wrong("Milk")
print(f"List 1: {list1}")

list2 = add_item_wrong("Bread")  # We expected just ["Bread"]!
print(f"List 2: {list2}")  # But we got ["Milk", "Bread"]

list3 = add_item_wrong("Eggs")
print(f"List 3: {list3}")  # ["Milk", "Bread", "Eggs"]!

# All lists are the SAME object!
print(f"\nAll same object? {list1 is list2 is list3}")
````

````python
# RIGHT WAY - Use None as default
def add_item_correct(item, shopping_list=None):
    """GOOD: Create new list if none provided."""
    if shopping_list is None:
        shopping_list = []
    shopping_list.append(item)
    return shopping_list

# Now each call gets its own list
list1 = add_item_correct("Milk")
print(f"List 1: {list1}")

list2 = add_item_correct("Bread")
print(f"List 2: {list2}")

# Or we can pass an existing list
my_list = ["Coffee", "Sugar"]
my_list = add_item_correct("Cream", my_list)
print(f"My list: {my_list}")
````

### *args: Variable Number of Positional Arguments

Sometimes you don't know how many arguments will be passed. Use `*args` to accept any number:

````python
def calculate_total(*amounts):
    """Calculate the sum of any number of amounts.
    
    The *args parameter collects all positional arguments into a tuple.
    """
    print(f"Received: {amounts}")  # It's a tuple
    print(f"Type: {type(amounts)}")
    return sum(amounts)

# Call with different numbers of arguments
print(f"Total: ${calculate_total(10, 20, 30):.2f}")
print()
print(f"Total: ${calculate_total(5.99, 12.50, 8.75, 22.00, 15.25):.2f}")
````

````python
# Practical example: Sales report
def generate_sales_summary(salesperson, *sales):
    """Generate a sales summary for a salesperson."""
    total = sum(sales)
    count = len(sales)
    average = total / count if count > 0 else 0
    
    print(f"\nSales Summary for {salesperson}")
    print(f"="*35)
    print(f"Number of sales: {count}")
    print(f"Total revenue: ${total:,.2f}")
    print(f"Average sale: ${average:,.2f}")
    if sales:
        print(f"Largest sale: ${max(sales):,.2f}")
        print(f"Smallest sale: ${min(sales):,.2f}")

generate_sales_summary("Alice", 1500, 2300, 890, 3200, 1750)
generate_sales_summary("Bob", 5000, 7500, 12000)
````

### **kwargs: Variable Number of Keyword Arguments

Use `**kwargs` to accept any number of keyword arguments:

````python
def create_product(**attributes):
    """Create a product dictionary from keyword arguments.
    
    The **kwargs parameter collects keyword arguments into a dictionary.
    """
    print(f"Received: {attributes}")
    print(f"Type: {type(attributes)}")
    return attributes

# Create products with different attributes
laptop = create_product(name="Laptop Pro", price=1299.99, brand="TechCorp", ram="16GB")
print()
shirt = create_product(name="Business Shirt", price=49.99, size="L", color="Blue")
````

````python
# Practical example: Configurable report generator
def generate_report(title, data, **options):
    """
    Generate a report with customizable options.
    
    Options can include:
        - show_header: bool (default True)
        - currency_symbol: str (default '$')
        - decimal_places: int (default 2)
        - width: int (default 40)
    """
    # Get options with defaults
    show_header = options.get('show_header', True)
    currency = options.get('currency_symbol', '$')
    decimals = options.get('decimal_places', 2)
    width = options.get('width', 40)
    
    if show_header:
        print("=" * width)
        print(f"{title:^{width}}")
        print("=" * width)
    
    for item, value in data.items():
        formatted_value = f"{currency}{value:,.{decimals}f}"
        print(f"{item:<20} {formatted_value:>{width-21}}")
    
    print("-" * width)

# Use with different options
sales_data = {"Q1": 125000, "Q2": 142000, "Q3": 138000, "Q4": 165000}

print("Default options:")
generate_report("Annual Sales Report", sales_data)

print("\nCustom options:")
generate_report("European Sales", sales_data, 
                currency_symbol='€', 
                width=50,
                decimal_places=0)
````

## 4.5 Variable Scope: Where Variables Live

**Scope** determines where a variable can be accessed. Understanding scope prevents bugs and confusion.

### Local Scope

Variables created inside a function are **local**—they exist only within that function:

````python
def calculate_bonus(salary):
    """Calculate employee bonus."""
    # These variables are LOCAL to this function
    bonus_rate = 0.10
    bonus = salary * bonus_rate
    print(f"Inside function: bonus_rate = {bonus_rate}")
    return bonus

result = calculate_bonus(50000)
print(f"Bonus calculated: ${result:,.2f}")

# Trying to access local variables outside the function causes an error
# Uncomment the line below to see the error:
# print(bonus_rate)  # NameError: name 'bonus_rate' is not defined
````

### Global Scope

Variables created outside functions are **global**—accessible from anywhere in the module:

````python
# Global variables
COMPANY_NAME = "TechCorp Industries"
TAX_RATE = 0.07  # Using UPPERCASE for constants is a convention

def print_invoice_header():
    """Print invoice header using global constant."""
    print(f"="*40)
    print(f"{COMPANY_NAME:^40}")  # Can read global variable
    print(f"="*40)

def calculate_with_tax(amount):
    """Calculate total using global tax rate."""
    return amount * (1 + TAX_RATE)  # Can read global variable

print_invoice_header()
print(f"Subtotal: $100.00")
print(f"Tax Rate: {TAX_RATE*100}%")
print(f"Total: ${calculate_with_tax(100):.2f}")
````

### The `global` Keyword: Modifying Global Variables

To modify (not just read) a global variable inside a function, you must declare it with `global`:

````python
# Global counter
transaction_count = 0
total_revenue = 0

def process_sale(amount):
    """Process a sale and update global counters."""
    global transaction_count, total_revenue  # Declare we're modifying globals
    
    transaction_count += 1
    total_revenue += amount
    print(f"Sale #{transaction_count}: ${amount:.2f}")

# Process some sales
process_sale(125.50)
process_sale(89.99)
process_sale(245.00)
process_sale(32.50)

print(f"\nTotal transactions: {transaction_count}")
print(f"Total revenue: ${total_revenue:.2f}")
````

### Best Practice: Avoid Modifying Globals

While the `global` keyword exists, it's generally better to return values from functions instead of modifying global state. Global modifications make code harder to understand and debug.

````python
# Better approach: Return values instead of modifying globals
def process_sales_batch(sales_list, current_count=0, current_total=0):
    """
    Process a batch of sales and return updated totals.
    
    This approach is better because:
    1. No global state to track
    2. Function is pure - same input always gives same output
    3. Easy to test
    4. Can be used in parallel processing
    """
    for amount in sales_list:
        current_count += 1
        current_total += amount
        print(f"Sale #{current_count}: ${amount:.2f}")
    
    return current_count, current_total

# Process sales and capture the results
sales = [125.50, 89.99, 245.00, 32.50]
count, total = process_sales_batch(sales)

print(f"\nTotal transactions: {count}")
print(f"Total revenue: ${total:.2f}")

# Process more sales, starting from previous totals
print("\nProcessing afternoon batch:")
afternoon_sales = [199.00, 55.75]
count, total = process_sales_batch(afternoon_sales, count, total)

print(f"\nFinal totals: {count} transactions, ${total:.2f} revenue")
````

### Scope Resolution: LEGB Rule

Python looks for variables in this order:

1. **L**ocal - Inside the current function
2. **E**nclosing - In outer functions (for nested functions)
3. **G**lobal - At the module level
4. **B**uilt-in - Python's built-in names (like `print`, `len`)

````python
# Demonstrating LEGB rule
x = "global"  # Global scope

def outer():
    x = "enclosing"  # Enclosing scope
    
    def inner():
        x = "local"  # Local scope
        print(f"inner sees: {x}")
    
    inner()
    print(f"outer sees: {x}")

outer()
print(f"global sees: {x}")
````

## 4.6 Built-in Functions: Python's Toolbox

Python provides many useful functions out of the box. Knowing these saves time and code.

### Numeric Functions

````python
# Sample data: Daily sales figures
daily_sales = [2340, 1890, 2780, 3150, 2090, 1650, 2920]

print("Daily Sales Analysis")
print("="*35)
print(f"Days recorded: {len(daily_sales)}")
print(f"Total sales: ${sum(daily_sales):,}")
print(f"Minimum day: ${min(daily_sales):,}")
print(f"Maximum day: ${max(daily_sales):,}")
print(f"Average: ${sum(daily_sales)/len(daily_sales):,.2f}")
````

### Sorting Functions

````python
# sorted() returns a new sorted list
sales_reps = ["Charlie", "Alice", "Bob", "Diana"]
monthly_sales = [45000, 62000, 51000, 58000]

print("Original order:", sales_reps)
print("Sorted alphabetically:", sorted(sales_reps))
print("Original unchanged:", sales_reps)  # sorted() doesn't modify original

print("\nSales low to high:", sorted(monthly_sales))
print("Sales high to low:", sorted(monthly_sales, reverse=True))
````

### When to Use Lambda vs Regular Functions

| Use Lambda When... | Use Regular Function When... |
|-------------------|-----------------------------|
| Simple, one-line operation | Multiple lines needed |
| Used only once (inline) | Reused in multiple places |
| With map(), filter(), sorted() | Need docstring/documentation |
| Short and readable | Complex logic |

**Rule of thumb:** If you need to assign a lambda to a variable and use it multiple times, write a regular function instead.

## 4.8 Docstrings: Documenting Your Functions

Good documentation makes your code usable by others (including future you!). Python has a built-in system for documentation called **docstrings**.

````python
def calculate_compound_interest(principal, rate, years, compounds_per_year=12):
    """
    Calculate the future value of an investment with compound interest.
    
    This function uses the compound interest formula:
    A = P(1 + r/n)^(nt)
    
    Args:
        principal (float): The initial investment amount.
        rate (float): Annual interest rate as a decimal (e.g., 0.05 for 5%).
        years (int): Number of years to invest.
        compounds_per_year (int, optional): Number of times interest compounds 
            per year. Defaults to 12 (monthly).
    
    Returns:
        float: The final investment value after compound interest.
    
    Examples:
        >>> calculate_compound_interest(10000, 0.05, 10)
        16470.09
        
        >>> calculate_compound_interest(5000, 0.08, 5, 4)  # Quarterly
        7429.74
    
    Raises:
        ValueError: If principal, rate, or years is negative.
    """
    if principal < 0 or rate < 0 or years < 0:
        raise ValueError("Principal, rate, and years must be non-negative")
    
    amount = principal * (1 + rate/compounds_per_year) ** (compounds_per_year * years)
    return round(amount, 2)

# Access the docstring
print(calculate_compound_interest.__doc__)
````

### Example 1: Financial Calculator Suite

````python
import math

def calculate_compound_interest(principal, annual_rate, years, compounds=12):
    """
    Calculate future value with compound interest.
    
    Formula: A = P(1 + r/n)^(nt)
    """
    amount = principal * (1 + annual_rate/compounds) ** (compounds * years)
    return amount

def calculate_monthly_payment(principal, annual_rate, years):
    """
    Calculate monthly loan payment.
    
    Formula: M = P * [r(1+r)^n] / [(1+r)^n - 1]
    """
    monthly_rate = annual_rate / 12
    num_payments = years * 12
    
    if monthly_rate == 0:
        return principal / num_payments
    
    payment = principal * (monthly_rate * (1 + monthly_rate)**num_payments) / \
              ((1 + monthly_rate)**num_payments - 1)
    return payment

def calculate_roi(initial_investment, final_value):
    """Calculate Return on Investment (ROI) as percentage."""
    roi = ((final_value - initial_investment) / initial_investment) * 100
    return roi

def calculate_break_even(fixed_costs, price_per_unit, variable_cost_per_unit):
    """Calculate break-even point in units."""
    if price_per_unit <= variable_cost_per_unit:
        return float('inf')  # Never breaks even
    return fixed_costs / (price_per_unit - variable_cost_per_unit)

# Use the financial calculator
print("="*50)
print("         FINANCIAL CALCULATOR SUITE")
print("="*50)

# Investment scenario
print("\n1. INVESTMENT GROWTH")
print("-"*50)
initial = 10000
rate = 0.07
years = 20

future_value = calculate_compound_interest(initial, rate, years)
roi = calculate_roi(initial, future_value)

print(f"Initial investment: ${initial:,.2f}")
print(f"Annual return: {rate*100}%")
print(f"Time period: {years} years")
print(f"Future value: ${future_value:,.2f}")
print(f"Total ROI: {roi:.1f}%")

# Loan scenario
print("\n2. LOAN PAYMENT")
print("-"*50)
loan_amount = 250000
interest_rate = 0.065
loan_term = 30

monthly = calculate_monthly_payment(loan_amount, interest_rate, loan_term)
total_paid = monthly * 12 * loan_term

print(f"Loan amount: ${loan_amount:,.2f}")
print(f"Interest rate: {interest_rate*100}%")
print(f"Term: {loan_term} years")
print(f"Monthly payment: ${monthly:,.2f}")
print(f"Total paid: ${total_paid:,.2f}")
print(f"Total interest: ${total_paid - loan_amount:,.2f}")

# Break-even analysis
print("\n3. BREAK-EVEN ANALYSIS")
print("-"*50)
fixed_costs = 50000
sale_price = 99.99
variable_cost = 45.00

break_even = calculate_break_even(fixed_costs, sale_price, variable_cost)

print(f"Fixed costs: ${fixed_costs:,.2f}")
print(f"Sale price: ${sale_price:.2f}")
print(f"Variable cost: ${variable_cost:.2f}")
print(f"Break-even point: {math.ceil(break_even):,} units")
print(f"Break-even revenue: ${break_even * sale_price:,.2f}")
````

### Example 2: Sales Commission System

````python
def calculate_commission(sales_amount, tier="standard", bonus_eligible=False):
    """
    Calculate sales commission based on tier and performance.
    
    Tiers:
        - standard: 5% base, 7% over $10K
        - senior: 7% base, 10% over $15K
        - executive: 10% flat rate
    
    Bonus: Additional 2% if bonus_eligible and sales >= $25K
    """
    tier_config = {
        "standard": {"base_rate": 0.05, "bonus_rate": 0.07, "threshold": 10000},
        "senior": {"base_rate": 0.07, "bonus_rate": 0.10, "threshold": 15000},
        "executive": {"base_rate": 0.10, "bonus_rate": 0.10, "threshold": 0}
    }
    
    config = tier_config.get(tier, tier_config["standard"])
    
    if sales_amount <= config["threshold"]:
        commission = sales_amount * config["base_rate"]
    else:
        commission = (config["threshold"] * config["base_rate"] + 
                     (sales_amount - config["threshold"]) * config["bonus_rate"])
    
    # Performance bonus
    if bonus_eligible and sales_amount >= 25000:
        commission += sales_amount * 0.02
    
    return commission

def generate_commission_report(sales_reps):
    """
    Generate commission report for all sales representatives.
    
    Args:
        sales_reps: List of dicts with name, sales, tier, bonus_eligible
    """
    print("\n" + "="*70)
    print("               MONTHLY COMMISSION REPORT")
    print("="*70)
    print(f"{'Name':<15} {'Tier':<12} {'Sales':>12} {'Commission':>12} {'Bonus':>8}")
    print("-"*70)
    
    total_sales = 0
    total_commission = 0
    
    for rep in sales_reps:
        commission = calculate_commission(
            rep["sales"], 
            rep["tier"], 
            rep.get("bonus_eligible", False)
        )
        total_sales += rep["sales"]
        total_commission += commission
        
        bonus_indicator = "Yes" if rep.get("bonus_eligible") and rep["sales"] >= 25000 else "-"
        
        print(f"{rep['name']:<15} {rep['tier']:<12} ${rep['sales']:>10,.2f} "
              f"${commission:>10,.2f} {bonus_indicator:>8}")
    
    print("-"*70)
    print(f"{'TOTALS':<28} ${total_sales:>10,.2f} ${total_commission:>10,.2f}")
    print("="*70)
    
    return total_sales, total_commission

# Sales team data
team = [
    {"name": "Alice Chen", "sales": 32000, "tier": "senior", "bonus_eligible": True},
    {"name": "Bob Martinez", "sales": 18500, "tier": "standard", "bonus_eligible": True},
    {"name": "Carol Johnson", "sales": 45000, "tier": "executive", "bonus_eligible": True},
    {"name": "David Kim", "sales": 8500, "tier": "standard", "bonus_eligible": False},
    {"name": "Eve Wilson", "sales": 27500, "tier": "senior", "bonus_eligible": True}
]

total_sales, total_commission = generate_commission_report(team)
print(f"\nCommission as % of Sales: {(total_commission/total_sales)*100:.2f}%")
````

### Example 3: Invoice Generator

````python
from datetime import datetime, timedelta
import random

def generate_invoice_number():
    """Generate a unique invoice number."""
    date_part = datetime.now().strftime("%Y%m%d")
    random_part = random.randint(1000, 9999)
    return f"INV-{date_part}-{random_part}"

def calculate_line_item(quantity, unit_price, discount_percent=0):
    """Calculate line item total with optional discount."""
    subtotal = quantity * unit_price
    discount = subtotal * (discount_percent / 100)
    return subtotal - discount

def generate_invoice(customer, items, tax_rate=0.07, payment_terms=30):
    """
    Generate a complete invoice.
    
    Args:
        customer: dict with name, address, email
        items: list of dicts with name, quantity, unit_price, discount
        tax_rate: tax rate as decimal
        payment_terms: days until payment due
    
    Returns:
        dict: Complete invoice data
    """
    invoice_num = generate_invoice_number()
    issue_date = datetime.now()
    due_date = issue_date + timedelta(days=payment_terms)
    
    # Calculate line items
    line_items = []
    subtotal = 0
    
    for item in items:
        line_total = calculate_line_item(
            item["quantity"],
            item["unit_price"],
            item.get("discount", 0)
        )
        line_items.append({
            "name": item["name"],
            "quantity": item["quantity"],
            "unit_price": item["unit_price"],
            "discount": item.get("discount", 0),
            "line_total": line_total
        })
        subtotal += line_total
    
    tax = subtotal * tax_rate
    total = subtotal + tax
    
    return {
        "invoice_number": invoice_num,
        "issue_date": issue_date,
        "due_date": due_date,
        "customer": customer,
        "line_items": line_items,
        "subtotal": subtotal,
        "tax_rate": tax_rate,
        "tax_amount": tax,
        "total": total,
        "payment_terms": payment_terms
    }

def print_invoice(invoice):
    """Display a formatted invoice."""
    width = 60
    
    print("\n" + "="*width)
    print(f"{'TECHCORP INDUSTRIES':^{width}}")
    print(f"{'123 Business Ave, Tampa, FL 33620':^{width}}")
    print("="*width)
    
    print(f"\nInvoice #: {invoice['invoice_number']}")
    print(f"Date: {invoice['issue_date'].strftime('%B %d, %Y')}")
    print(f"Due Date: {invoice['due_date'].strftime('%B %d, %Y')}")
    print(f"Terms: Net {invoice['payment_terms']} days")
    
    print(f"\nBill To:")
    print(f"  {invoice['customer']['name']}")
    print(f"  {invoice['customer']['address']}")
    print(f"  {invoice['customer']['email']}")
    
    print("\n" + "-"*width)
    print(f"{'Item':<25} {'Qty':>5} {'Price':>10} {'Disc':>6} {'Total':>10}")
    print("-"*width)
    
    for item in invoice['line_items']:
        disc_str = f"{item['discount']}%" if item['discount'] > 0 else "-"
        print(f"{item['name']:<25} {item['quantity']:>5} ${item['unit_price']:>8.2f} "
              f"{disc_str:>6} ${item['line_total']:>8.2f}")
    
    print("-"*width)
    print(f"{'Subtotal:':>48} ${invoice['subtotal']:>8.2f}")
    print(f"{'Tax (' + str(invoice['tax_rate']*100) + '%)':>48} ${invoice['tax_amount']:>8.2f}")
    print("="*width)
    print(f"{'TOTAL DUE:':>48} ${invoice['total']:>8.2f}")
    print("="*width)
    print(f"\n{'Thank you for your business!':^{width}}")

# Create and print an invoice
customer = {
    "name": "Acme Corporation",
    "address": "456 Commerce St, Orlando, FL 32801",
    "email": "accounts@acme.com"
}

order_items = [
    {"name": "Laptop Pro 15", "quantity": 3, "unit_price": 1299.99, "discount": 5},
    {"name": "Wireless Mouse", "quantity": 10, "unit_price": 49.99},
    {"name": "USB-C Hub", "quantity": 5, "unit_price": 79.99, "discount": 10},
    {"name": "Monitor Stand", "quantity": 3, "unit_price": 149.99}
]

invoice = generate_invoice(customer, order_items)
print_invoice(invoice)
````

## 4.11 Best Practices Summary

### Function Design Principles

| Principle | Description | Example |
|-----------|-------------|--------|
| **Single Responsibility** | Each function should do one thing well | `calculate_tax()`, not `calculate_and_print_tax()` |
| **Descriptive Names** | Names should explain what the function does | `validate_email()` not `check()` |
| **Reasonable Size** | Keep functions short and focused | If > 20-30 lines, consider splitting |
| **Document Purpose** | Use docstrings to explain parameters and returns | See docstring examples above |
| **Minimize Side Effects** | Prefer returning values over modifying globals | Return results, don't print inside |
| **Use Default Arguments** | Make functions flexible but easy to use | `def greet(name, greeting="Hello"):` |

### Common Mistakes to Avoid

1. **Mutable default arguments** - Use `None` instead of `[]` or `{}`
2. **Modifying global state** - Return values instead
3. **Too many parameters** - Consider using a dictionary or object
4. **No error handling** - Validate inputs for public functions
5. **Mixing concerns** - Don't calculate AND print in the same function
