## 9.1 Why Data Visualization?

**Visualization turns raw numbers into insight.** Humans process visuals 60,000 times faster than text — a well-chosen chart can communicate in seconds what a table of numbers cannot.

Think about it this way: if someone hands you a spreadsheet with 1,000 rows of sales data, you'd need to scroll through every row to spot trends. But a single line chart can show you the overall pattern instantly. That's the power of visualization.

Data visualization helps you:

- **Reveal patterns** hidden in raw data — trends, outliers, and clusters jump out visually
- **Make data accessible** to non-technical stakeholders who may not understand code or statistics
- **Support business reporting** and decision-making with clear, compelling evidence
- **Verify your own analysis** — does the data actually look the way your numbers suggest?

**If you can't visualize it, you can't communicate it.**

### Business Applications

Data visualization is used across every business function. Whether you're in marketing, finance, operations, or strategy, you'll need to present data visually. Here are some common applications:

- **Executive dashboards and KPI tracking** — showing at a glance whether the business is on target
- **Sales and performance reports** — comparing actual vs. planned, ranking performers
- **Financial analysis and forecasting** — visualizing revenue trends, cost breakdowns
- **Customer segmentation and insights** — understanding who your customers are
- **Trend analysis and anomaly detection** — spotting unusual patterns early

**The ability to create clear visualizations is one of the most marketable skills in business analytics.** Employers consistently rank data visualization as a top skill for analysts.

## 9.2 Getting Started with Matplotlib

**Matplotlib is the foundational plotting library in Python.** It was created to replicate MATLAB's plotting capabilities and has become the most widely used visualization library in the Python ecosystem. Libraries like Seaborn and even Pandas' built-in plotting are built on top of Matplotlib.

The `pyplot` module provides a simple, MATLAB-like interface for creating charts. You'll almost always use it alongside Pandas and NumPy.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# For Jupyter notebooks (renders plots inline instead of in a popup)
%matplotlib inline

print("Libraries imported and ready!")
````

**Convention:** Always import pyplot as `plt` — you'll see this in every data science tutorial, textbook, and Stack Overflow answer. It's as standard as importing NumPy as `np` or Pandas as `pd`.

### Your First Plot

The basic Matplotlib workflow has three steps:

1. **Create the plot** — call a plotting function like `plt.plot()`
2. **Add labels** — set the title, axis labels, and any other annotations
3. **Display it** — call `plt.show()` to render the chart

Every chart you make follows this pattern, regardless of chart type.

````python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [42, 48, 45, 55]

plt.figure(figsize=(8, 4))            # Set the figure size
plt.plot(quarters, revenue)            # Step 1: Create the plot
plt.title('Quarterly Revenue ($K)')    # Step 2: Add labels
plt.xlabel('Quarter')
plt.ylabel('Revenue ($K)')
plt.show()                             # Step 3: Display it
````

**Always add a title and axis labels.** A chart without labels is meaningless to your audience — they have no way to know what they're looking at. Even if it seems obvious to you, always label your charts.

### Choosing the Right Chart

**Picking the right chart type is the most important visualization decision.** The wrong chart can mislead your audience or hide your findings. Before you start coding, ask yourself: *what am I trying to show?*

| Chart | Best For | Example |
|-------|----------|---------|
| **Line** | Trends over time | Monthly sales |
| **Bar** | Comparing categories | Revenue by region |
| **Scatter** | Relationships between variables | Ad spend vs. sales |
| **Pie** | Proportions of a whole | Market share |
| **Histogram** | Distribution of values | Salary ranges |

A quick rule of thumb:
- If your x-axis is **time**, use a **line chart**
- If your x-axis is **categories**, use a **bar chart**
- If both axes are **numbers**, use a **scatter plot**
- If you're showing **parts of a whole**, use a **pie chart**
- If you want to see the **spread** of one variable, use a **histogram**

## 9.3 Line Charts

**Line charts are the go-to for showing trends over time.** They connect data points in order, making it easy to see patterns like growth, decline, or seasonality. Whenever your x-axis represents a time period (days, months, quarters, years), a line chart is usually the right choice.

````python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [120, 145, 132, 158]

plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', color='steelblue')
plt.title('Monthly Sales')
plt.ylabel('Sales ($K)')
plt.grid(True, alpha=0.3)
plt.show()
````

Let's break down what's new here:

- **`plt.figure(figsize=(8, 5))`** — Creates a figure that is 8 inches wide and 5 inches tall. Without this, Matplotlib uses a default size that may be too small or too large.
- **`marker='o'`** — Adds a circle at each data point. Without markers, you only see the line, and it's hard to tell exactly where each value falls.
- **`color='steelblue'`** — Sets the line color. Matplotlib supports many named colors like `'red'`, `'green'`, `'coral'`, `'steelblue'`, etc.
- **`plt.grid(True, alpha=0.3)`** — Adds a subtle background grid. The `alpha=0.3` makes it 30% opaque (semi-transparent) so it doesn't overpower the data.

### Line Chart Customization

**Matplotlib gives you full control over line appearance.** You can set color, style, width, and labels for each line. Here are the key parameters you can pass to `plt.plot()`:

````python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [120, 145, 132, 158]

plt.figure(figsize=(8, 5))
plt.plot(months, sales,
         marker='o',       # Circle markers at each data point
         linestyle='-',    # Solid line (other options: '--', ':', '-.')
         color='blue',     # Line color
         linewidth=2,      # Thicker line (default is 1)
         label='2024')     # Label for the legend

plt.title('Monthly Sales')
plt.ylabel('Sales ($K)')
plt.legend()               # Display the legend using the label above
plt.grid(True, alpha=0.3)
plt.show()
````

**The `label` parameter and `plt.legend()` work together.** When you set `label='2024'` in the plot call, that text appears in the legend. But the legend only shows up if you also call `plt.legend()`. This becomes essential when you have multiple lines on the same chart.

### Multiple Lines

**Comparing datasets on one chart is as simple as calling `plt.plot()` multiple times.** Each call adds a new line to the same figure. Matplotlib automatically assigns different colors, but you should add labels and a legend so readers can tell the lines apart.

````python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales_2024 = [120, 145, 132, 158]
sales_2025 = [135, 160, 148, 175]

plt.figure(figsize=(8, 5))
plt.plot(months, sales_2024, marker='o', label='2024')
plt.plot(months, sales_2025, marker='s', label='2025')  # 's' = square marker
plt.title('Sales Comparison: 2024 vs 2025')
plt.ylabel('Sales ($K)')
plt.legend()
plt.show()
````

Notice that each `plt.plot()` call adds a new line to the same figure until you call `plt.show()`. At that point, the figure is rendered and a new blank figure starts for the next plot.

**Common marker styles:**
- `'o'` — circle
- `'s'` — square
- `'^'` — triangle up
- `'d'` — diamond
- `'*'` — star

Using different markers for each line helps readers distinguish them even in black-and-white printouts.

## 9.4 Bar Charts

**Bar charts are ideal for comparing values across categories.** They make it easy to see which category is largest, smallest, or how they rank relative to each other. Unlike line charts (which imply a continuous connection between points), bar charts emphasize that each category is distinct.

````python
regions = ['North', 'South', 'East', 'West']
revenue = [45000, 38000, 52000, 41000]

plt.figure(figsize=(8, 5))
plt.bar(regions, revenue, color='steelblue')
plt.title('Revenue by Region')
plt.ylabel('Revenue ($)')
plt.show()
````

The `plt.bar()` function takes two main arguments: the category names (x-axis) and the values (heights of the bars). The `color` parameter works the same way as with line charts.

### Bar Chart with Value Labels

**Adding value labels directly on bars makes charts self-explanatory.** Without labels, readers have to visually estimate values by looking across to the y-axis, which is imprecise. With labels, the exact value is right there on the bar.

This technique uses `plt.text()` to place text at specific coordinates on the chart. We loop through each bar object to get its position and height.

````python
regions = ['North', 'South', 'East', 'West']
revenue = [45000, 38000, 52000, 41000]

bars = plt.bar(regions, revenue, color='steelblue')

# Add value labels on top of each bar
for bar, val in zip(bars, revenue):
    plt.text(bar.get_x() + bar.get_width()/2,  # Center horizontally on the bar
             bar.get_height() + 500,             # Slightly above the top of the bar
             f'${val:,}', ha='center')           # Formatted with $ and commas

plt.title('Revenue by Region')
plt.show()
````

Let's break down the `plt.text()` call:

- **`bar.get_x() + bar.get_width()/2`** — This calculates the center of the bar horizontally. `get_x()` gives the left edge, and adding half the width gives the center.
- **`bar.get_height() + 500`** — This positions the text slightly above the top of the bar (500 units of padding).
- **`f'${val:,}'`** — An f-string that formats the number with a dollar sign and commas (e.g., `$45,000`).
- **`ha='center'`** — Horizontal alignment. This centers the text over the x position we calculated.

### Horizontal Bar Chart

**Use horizontal bars when category labels are long.** When labels are displayed on the y-axis (horizontally), they're much easier to read than when they're crammed under vertical bars. Horizontal bars are also great for ranking-style charts.

````python
regions = ['North', 'South', 'East', 'West']
revenue = [45000, 38000, 52000, 41000]

plt.barh(regions, revenue, color='steelblue')
plt.xlabel('Revenue ($)')
plt.title('Revenue by Region')
plt.show()
````

**Important:** `barh()` draws bars from bottom to top — the first item in your list appears at the bottom, and the last item appears at the top. So if you want the highest value at the top (which readers expect), sort your data in ascending order. The last item in an ascending-sorted list will be the largest and will appear at the top of the chart.

### Grouped Bar Chart

**Grouped bars let you compare multiple metrics side by side for each category.** For example, you might want to compare Q1 vs. Q2 sales for each region.

The trick is using `np.arange()` to create numeric x-positions, then offsetting each group of bars by the bar width. Without this offset, the bars would overlap.

````python
regions = ['North', 'South', 'East', 'West']
q1_sales = [120, 98, 140, 110]
q2_sales = [135, 115, 155, 125]

x = np.arange(len(regions))  # [0, 1, 2, 3] — numeric positions for each region
width = 0.35                  # Width of each bar

plt.bar(x - width/2, q1_sales, width, label='Q1')  # Shift left by half a bar width
plt.bar(x + width/2, q2_sales, width, label='Q2')  # Shift right by half a bar width
plt.xticks(x, regions)  # Replace numeric positions with region names
plt.title('Quarterly Sales by Region')
plt.legend()
plt.show()
````

Here's how the positioning works step by step:

1. `np.arange(len(regions))` creates `[0, 1, 2, 3]` — one position per region
2. `x - width/2` shifts Q1 bars slightly left of center: `[-0.175, 0.825, 1.825, 2.825]`
3. `x + width/2` shifts Q2 bars slightly right of center: `[0.175, 1.175, 2.175, 3.175]`
4. `plt.xticks(x, regions)` puts the region names at the center of each pair

This pattern works for any number of groups — just adjust the width and offsets accordingly.

## 9.5 Scatter Plots

**Scatter plots reveal relationships between two numerical variables.** Each point represents one observation, and the overall pattern of points shows whether a correlation exists. Unlike line charts, scatter plots don't connect the dots — they let you see the raw data distribution.

When you see points trending upward from left to right, that's a **positive correlation** (as one variable increases, so does the other). When they trend downward, that's a **negative correlation**.

````python
np.random.seed(7)
advertising = np.random.randint(50, 650, 40)
sales = advertising * 0.12 + np.random.normal(0, 8, 40)

plt.figure(figsize=(8, 5))
plt.scatter(advertising, sales,
            color='blue', alpha=0.6)
plt.xlabel('Advertising ($)')
plt.ylabel('Sales ($K)')
plt.title('Advertising vs Sales')
plt.show()
````

**`alpha=0.6` controls transparency.** An alpha of 1.0 is fully opaque; 0.0 is fully transparent. Using transparency helps in two ways:

1. When points overlap, darker areas indicate where multiple points cluster
2. It makes the chart look less cluttered

We use `np.random.seed(7)` so the random data is reproducible — everyone gets the same "random" points. The data simulates a linear relationship: `sales ≈ 0.12 × advertising + noise`.

### Scatter — Color by a Third Variable

**Use the `c=` parameter to color each point by a third variable.** This turns a 2D scatter plot into a 3D visualization — the color encodes additional information. The `cmap` parameter sets the color scheme (e.g., `'viridis'` goes from purple to yellow, where lighter colors = higher values).

````python
np.random.seed(10)
ad_spend = np.random.randint(100, 600, 20)
sales = ad_spend * 0.12 + np.random.normal(0, 5, 20)
satisfaction = np.random.uniform(3.0, 5.0, 20)  # Third variable

plt.figure(figsize=(8, 5))
scatter = plt.scatter(ad_spend, sales, c=satisfaction,
                      cmap='viridis', alpha=0.8, edgecolors='gray')
plt.colorbar(scatter, label='Customer Satisfaction')
plt.xlabel('Ad Spend ($)')
plt.ylabel('Sales ($K)')
plt.title('Ad Spend vs Sales (colored by Satisfaction)')
plt.show()
````

### Scatter with Trend Line

**Adding a trend line helps quantify the relationship** you see in the scatter pattern. A trend line (also called a regression line or line of best fit) is a straight line that best represents the overall direction of the data.

We use two NumPy functions to create the trend line:
- **`np.polyfit(x, y, 1)`** — Finds the slope and intercept of the best-fit line (the `1` means first-degree polynomial, i.e., a straight line)
- **`np.poly1d(z)`** — Turns those coefficients into a callable function you can use to calculate y-values

````python
# np.array() is needed for polyfit/poly1d
advertising = np.array([100, 200, 300, 400, 500, 600])
sales = np.array([22, 35, 48, 58, 72, 80])

plt.scatter(advertising, sales, alpha=0.7)

# Fit a line: polyfit returns slope & intercept, poly1d makes it callable
z = np.polyfit(advertising, sales, 1)
trend = np.poly1d(z)
plt.plot(advertising, trend(advertising), "r--", label='Trend')
plt.xlabel('Advertising ($)')
plt.ylabel('Sales ($K)')
plt.title('Ad Spend vs Sales with Trend')
plt.legend()
plt.show()
````

Let's trace through the trend line code:

1. **`np.polyfit(advertising, sales, 1)`** returns an array like `[0.116, 9.5]`, meaning slope ≈ 0.116 and intercept ≈ 9.5
2. **`np.poly1d(z)`** creates a function: `trend(x) = 0.116x + 9.5`
3. **`trend(advertising)`** calculates the y-value on the line for each x-value
4. **`"r--"`** is a shorthand format string: `r` = red, `--` = dashed line

**Important:** The data must be NumPy arrays (not plain Python lists) for `polyfit` and `poly1d` to work. That's why we wrap them with `np.array()` at the top.

**A dashed red trend line (`"r--"`) is a common convention** for showing the overall direction of data.

## 9.6 Pie Charts and Histograms

### Pie Charts

**Pie charts show proportions of a whole.** Each slice represents a category's share of the total. They work best with 3-5 categories — more than that and the slices become too thin to distinguish.

**When to use pie charts:** When you want to emphasize that parts add up to 100% — for example, market share, budget allocation, or revenue breakdown by product line.

````python
sizes = [35, 25, 20, 12, 8]
labels = ['Product A', 'Product B', 'Product C',
          'Product D', 'Product E']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')  # Shows "35.0%"
plt.title('Market Share')
plt.show()
````

**Understanding `autopct='%1.1f%%'`:** This is a C-style format string that tells Matplotlib how to display the percentage on each slice.

- `%` — Start of the format code
- `1.1f` — Display as a float with 1 decimal place (e.g., `35.0`)
- `%%` — A literal percent sign (`%`)

So a slice representing 35% of the total displays as `35.0%`. If you want no decimals, use `'%1.0f%%'` instead (displays `35%`).

Matplotlib automatically calculates the percentages for you based on the values you pass in — you don't need to ensure they sum to 100.

### Pie Chart Options

**You can emphasize a slice with `explode` and add depth with `shadow`.** You can also use a custom color palette to match your brand or report style.

````python
sizes = [35, 25, 20, 12, 8]
labels = ['Product A', 'Product B', 'Product C',
          'Product D', 'Product E']
colors = ['#3498db', '#e74c3c', '#2ecc71',
          '#f39c12', '#9b59b6']

plt.pie(sizes, labels=labels, autopct='%1.0f%%',
        explode=(0.05, 0, 0, 0, 0),  # Pull out first slice slightly
        shadow=True, colors=colors)   # Shows "35%"
plt.title('Market Share')
plt.show()
````

The `explode` parameter is a tuple with one value per slice. Each value is the fraction of the pie radius to offset that slice:
- `0` — No offset (flush with the rest of the pie)
- `0.05` — Pull out by 5% of the radius (subtle emphasis)
- `0.1` — Pull out by 10% (more dramatic)

**Tip:** Only `explode` one slice to draw attention to it — exploding all of them defeats the purpose and makes the chart harder to read.

### Histogram vs Bar Chart

**These look similar but serve different purposes.** Students often confuse them — here's the key distinction:

| | Bar Chart | Histogram |
|---|-----------|-----------|
| **Data** | Categories (regions, products) | Continuous numbers (salaries, ages) |
| **X-axis** | Distinct labels | Ranges (bins) |
| **Bars** | Separated (gaps between) | Touching (continuous ranges) |
| **Question** | "How do categories compare?" | "How is data spread out?" |

**Bar chart:** "Which region has the most revenue?" **Histogram:** "What salary range is most common?"

### Histograms

**Histograms show the distribution (spread) of a single variable.** While bar charts compare categories, histograms group continuous values into ranges called **bins** and show how frequently values fall into each range.

For example, a histogram of employee salaries might show that most salaries cluster around $60,000-$70,000, with fewer at the extremes. This kind of shape analysis is impossible with a bar chart or line chart.

````python
# Generate 500 random salaries from a normal distribution
# Mean = $65,000, Standard deviation = $15,000
salaries = np.random.normal(65000, 15000, 500)

plt.hist(salaries, bins=20,
         color='steelblue',
         edgecolor='black')   # Black edges make bins easier to distinguish
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')       # How many values fall in each bin
plt.title('Salary Distribution')
plt.show()
````

**Choosing the right number of `bins`:**

- **Too few bins** (e.g., 5) hides the shape of the distribution — everything gets lumped together
- **Too many bins** (e.g., 100) creates noise — the chart looks jagged and hard to interpret
- **A good starting point** is 15-25 bins, then adjust based on what you see

The `edgecolor='black'` parameter adds a black border around each bar, making it easier to see where one bin ends and the next begins. Without it, adjacent bars of the same color blend together.

## 9.7 Annotations and Reference Lines

**Annotations and reference lines help highlight key values on a chart.** They transform a basic chart into a story by drawing attention to important data points, thresholds, or targets.

Matplotlib provides three tools for this:
- **`plt.axhline()`** — Draws a horizontal line across the entire chart (e.g., a target or average)
- **`plt.axvline()`** — Draws a vertical line across the entire chart (e.g., marking a specific date)
- **`plt.annotate()`** — Adds a text callout with an arrow pointing to a specific data point

````python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [120, 145, 132, 158]

plt.plot(months, sales, marker='o')
plt.title('Monthly Sales with Targets')

# Horizontal reference line (e.g., a sales target)
plt.axhline(y=140, color='red', linestyle='--', label='Target')

# Vertical reference line (e.g., mark a specific month)
plt.axvline(x='Mar', color='gray', linestyle=':', alpha=0.5)

# Annotate a specific point with an arrow
plt.annotate('Peak!', xy=('Apr', 158),       # Where the arrow points TO
             xytext=('Mar', 165),              # Where the text sits
             arrowprops=dict(arrowstyle='->'),  # Arrow style
             fontsize=12, fontweight='bold')
plt.legend()
plt.show()
````

The `plt.annotate()` function has two position parameters:

- **`xy`** — The point the arrow should point **to** (the data point you're annotating)
- **`xytext`** — Where the text label should be placed (typically offset from the data point so it doesn't overlap)

The `arrowprops` parameter controls the arrow's appearance. `arrowstyle='->'` gives a simple arrow. Other options include `'-|>'` (filled arrowhead) and `'fancy'` (curved arrow).

**Common reference line uses:**
- Target/goal lines ("We need to hit $140K")
- Average lines ("The mean is here")
- Date markers ("This is when we launched the campaign")
- Threshold lines ("Anything below this is a loss")

## 9.8 Subplots and Dashboards

**Subplots let you put multiple plots in one figure** — essential for dashboards and comparative analysis. Instead of showing four separate charts, you can combine them into a single figure where the reader sees everything at once.

`plt.subplots()` returns two objects:
- **`fig`** — The entire figure (the canvas that contains all the plots)
- **`axes`** — A grid of individual plot areas that you draw on

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# axes[0] = left plot, axes[1] = right plot

# Left plot
axes[0].plot([1, 2, 3], [10, 20, 15], marker='o')
axes[0].set_title('Sales Trend')

# Right plot
axes[1].bar(['A', 'B', 'C'], [300, 450, 200])
axes[1].set_title('Revenue by Product')

plt.tight_layout()  # Prevents labels from overlapping between subplots
plt.show()
````

### Important: Subplot Method Names

**With subplots, you use `axes[i]` methods instead of `plt.` functions.** The method names are slightly different:

| Single plot (`plt.`) | Subplot (`axes[i].`) |
|---------------------|---------------------|
| `plt.title()` | `axes[i].set_title()` |
| `plt.xlabel()` | `axes[i].set_xlabel()` |
| `plt.ylabel()` | `axes[i].set_ylabel()` |
| `plt.legend()` | `axes[i].legend()` |
| `plt.axhline()` | `axes[i].axhline()` |

Notice the pattern: single-plot functions like `plt.title()` become `axes[i].set_title()` (with the `set_` prefix). Functions like `plt.legend()` and `plt.axhline()` keep the same name.

**`plt.tight_layout()`** is important whenever you use subplots — it automatically adjusts the spacing so titles and labels don't overlap.

### Subplots (2x2 Grid)

**For a 2x2 grid, `axes` becomes a 2D array** — you index it with `[row, col]`. This is the foundation for creating dashboards.

````python
x, y = [1, 2, 3, 4], [10, 25, 18, 30]
cats = ['A', 'B', 'C', 'D']
sizes = [40, 30, 20, 10]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(x, y, marker='o')     # Top left
axes[0, 1].bar(cats, y)               # Top right
axes[1, 0].scatter(x, y)              # Bottom left
axes[1, 1].pie(sizes, labels=cats)    # Bottom right

fig.suptitle('My Dashboard', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````

Here's how the 2D indexing works:

```
axes[0, 0] = Top left      axes[0, 1] = Top right
axes[1, 0] = Bottom left   axes[1, 1] = Bottom right
```

**`fig.suptitle()`** adds a main title that spans the entire figure, above all the subplots. This is different from `axes[i].set_title()`, which only titles one subplot. Use `fig.suptitle()` for the overall dashboard title, and `set_title()` for individual chart titles.

### Business Dashboard Example

**Putting it all together:** A multi-panel dashboard combining different chart types to tell a complete story. Each subplot shows a different aspect of the data, with proper titles, labels, colors, and a trend line.

````python
months = ['Oct', 'Nov', 'Dec']
revenue, profit = [150, 165, 180], [45, 38, 52]
regions = ['North', 'South', 'East']
shares = [40, 35, 25]
spend = np.array([100, 200, 300, 400])
sales = np.array([22, 35, 48, 58])

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Q4 Performance Dashboard', fontsize=16, fontweight='bold')

axes[0, 0].plot(months, revenue, marker='o', color='steelblue')
axes[0, 0].set_title('Revenue Trend')
axes[0, 0].set_ylabel('Revenue ($K)')

axes[0, 1].bar(regions, profit, color=['#3498db', '#e74c3c', '#2ecc71'])
axes[0, 1].set_title('Profit by Region')
axes[0, 1].set_ylabel('Profit ($K)')

axes[1, 0].pie(shares, labels=regions, autopct='%1.0f%%',
               colors=['#3498db', '#e74c3c', '#2ecc71'])
axes[1, 0].set_title('Regional Market Share')

axes[1, 1].scatter(spend, sales, alpha=0.7, color='steelblue')
z = np.polyfit(spend, sales, 1)
axes[1, 1].plot(spend, np.poly1d(z)(spend), 'r--')
axes[1, 1].set_title('Ad Spend vs Sales')
axes[1, 1].set_xlabel('Spend ($)')
axes[1, 1].set_ylabel('Sales ($K)')

plt.tight_layout()
plt.savefig('dashboard.png', dpi=300, bbox_inches='tight')
plt.show()
````

Notice that this example uses `plt.savefig()` *before* `plt.show()`. This is important — `plt.show()` clears the figure from memory, so if you try to save after showing, you'll get a blank image. We'll cover saving in more detail in Section 9.10.

## 9.10 Customization and Saving

### Figure Size and Layout

**Use `plt.figure()` for single charts and `plt.subplots()` when you need multiple plots.** Both accept `figsize=(width, height)` in inches.

````python
# Standalone figure
plt.figure(figsize=(10, 6))
plt.plot([1, 2, 3], [10, 20, 15])
plt.title('Single Chart with figsize=(10, 6)')
plt.show()

# Or with subplots (same figsize parameter)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot([1, 2, 3], [10, 20, 15])
ax.set_title('Same size, using subplots')
plt.show()
````

**Standard sizes to remember:**
- `(10, 6)` — Good for single charts in reports
- `(8, 5)` — Compact single chart
- `(12, 10)` or `(14, 10)` — Dashboards with subplots

The size is in inches, which matters when you save the figure — a `(10, 6)` figure saved at 300 DPI produces a 3000×1800 pixel image.

### Customization: Before and After

**Professional charts need consistent formatting.** Here's the same data presented two ways — bare defaults vs. polished. Notice how a few extra lines of code make the chart dramatically more readable:

````python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_a = [42, 48, 45, 55, 60, 58]
sales_b = [35, 40, 50, 47, 52, 61]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# BEFORE: bare defaults
axes[0].plot(months, sales_a)
axes[0].plot(months, sales_b)
axes[0].set_title('Before')

# AFTER: polished with markers, labels, legend, grid, fonts
axes[1].plot(months, sales_a, marker='o', label='Product A')
axes[1].plot(months, sales_b, marker='s', label='Product B')
axes[1].set_title('After', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Sales ($K)', fontsize=12)
axes[1].legend(loc='upper left')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

**Key additions that make the "After" chart professional:**

- **`marker='o'`** / **`marker='s'`** — Shows individual data points
- **`label=` + `legend()`** — Tells readers which line is which
- **`fontsize=14, fontweight='bold'`** — Makes the title stand out
- **`grid(True, alpha=0.3)`** — Adds subtle gridlines for reference
- **`set_ylabel()`** — Labels the axis with units

These six additions take about 30 seconds to type and transform a chart from "rough draft" to "presentation-ready."

### Color Options

**Matplotlib supports named colors, hex codes, and color palettes.** Consistent colors across your charts make reports look professional. Here's a cohesive example using a brand color palette:

````python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [42, 55, 48, 63]

# Define a brand color palette using hex codes
COLORS = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']

plt.figure(figsize=(8, 5))
plt.bar(quarters, revenue, color=COLORS)
plt.title('Quarterly Revenue', fontsize=14, fontweight='bold')
plt.ylabel('Revenue ($K)')
plt.show()

# Three ways to specify colors:
# 1. Named colors: color='steelblue', color='coral'
# 2. Hex codes:    color='#3498db'  (precise, for brand consistency)
# 3. Color list:   color=['#3498db', '#e74c3c', ...]  (one per bar)
````

**Hex colors** are six-digit codes that precisely specify a color (like `#3498db` for a specific shade of blue). You've probably seen these in web design or presentation software. They're useful when you need exact brand consistency — for example, USF's green is `#006747`.

Some commonly used named colors: `'steelblue'`, `'coral'`, `'green'`, `'orange'`, `'purple'`, `'gray'`.

### Saving Figures

**After creating a chart, you can save it as an image file** for use in reports, presentations, or websites. Matplotlib supports many formats including PNG, PDF, and SVG.

````python
# Create a chart
plt.figure(figsize=(8, 5))
plt.bar(['A', 'B', 'C'], [10, 20, 15], color='steelblue')
plt.title('Sample Chart')

# Save as PNG (for web/presentations) — BEFORE plt.show()!
# plt.savefig('chart.png', dpi=300, bbox_inches='tight')

# Save as PDF (for print)
# plt.savefig('chart.pdf', bbox_inches='tight')

plt.show()

print("Key savefig parameters:")
print("  dpi=300           — Resolution (dots per inch). 300 is print-quality.")
print("  bbox_inches='tight' — Removes extra whitespace around the chart.")
````

**Critical rule: Call `savefig()` BEFORE `plt.show()`.** When Matplotlib displays a chart with `show()`, it clears the figure from memory. If you call `savefig()` after `show()`, you'll save a blank image.

**Always use `bbox_inches='tight'`** — without it, axis labels and titles can get cut off at the edges of the saved image.

**When to use which format:**
- **PNG** — Best for web, presentations, and screen display. Use `dpi=300` for high quality.
- **PDF** — Best for print reports and documents. Vector format, so it scales perfectly.

## 9.11 Best Practices

Creating a chart is easy. Creating a *good* chart takes thought. Here are five principles that will make your visualizations effective:

1. **Choose the right chart** for your data type and message. Don't use a pie chart when a bar chart would be clearer. Don't use a scatter plot when you have time-series data.

2. **Label everything** — title, axes, legend, units. Your audience should be able to understand the chart without any additional explanation. Include units (dollars, percentages, thousands) so there's no ambiguity.

3. **Use color purposefully** — highlight key data, don't decorate. Color should draw attention to the most important information. Using too many colors makes everything compete for attention.

4. **Keep it simple** — remove chart junk, let the data speak. Don't add 3D effects, excessive gridlines, or decorative elements that don't convey information. Every element on your chart should serve a purpose.

5. **Consider your audience** — executives want summaries, analysts want detail. A CEO wants to see the one key number and whether it's good or bad. An analyst wants to see the full distribution and methodology. Design your chart for its intended audience.
