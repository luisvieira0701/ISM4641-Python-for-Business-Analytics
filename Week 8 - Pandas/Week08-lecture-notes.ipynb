## 8.1 Understanding Pandas and Its Ecosystem

### What Makes Pandas Special?

Pandas brings together the best aspects of several data manipulation paradigms:

| Feature | Pandas Advantage |
|---------|------------------|
| **Labeled Data** | Unlike NumPy arrays, Pandas structures have named columns and row labels |
| **Heterogeneous Data** | Columns can have different data types (numbers, strings, dates) |
| **Missing Data Handling** | Built-in support for NaN (Not a Number) values |
| **Data Alignment** | Automatic alignment of data based on labels |
| **Rich I/O** | Read/write CSV, Excel, SQL, JSON, Parquet, and more |
| **Time Series** | First-class support for date/time data |

### The Pandas Relationship with NumPy

Pandas is built on top of NumPy. Every Pandas object contains NumPy arrays underneath. This means:
- Pandas inherits NumPy's speed advantages
- NumPy functions work with Pandas objects
- You can extract NumPy arrays from Pandas when needed

````python
# Import pandas - always use the alias 'pd'
import pandas as pd
import numpy as np

print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")

# Pandas uses NumPy under the hood
sample_series = pd.Series([1, 2, 3, 4, 5])
print(f"\nUnderlying NumPy array: {sample_series.values}")
print(f"Type: {type(sample_series.values)}")
````

## 8.2 Series: The One-Dimensional Building Block

A **Series** is a one-dimensional labeled array capable of holding any data type. Think of it as a single column from a spreadsheet, but with superpowers.

### Anatomy of a Series

Every Series has two components:
1. **Values**: The actual data (stored as a NumPy array)
2. **Index**: Labels for each value (default is 0, 1, 2, ...)

```
Index    Values
  0  ->   250
  1  ->   180
  2  ->   320
  3  ->   290
  4  ->   410
```

````python
# Creating a Series from a list (default integer index)
daily_sales = pd.Series([250, 180, 320, 290, 410])
print("Daily Sales Series:")
print(daily_sales)
print(f"\nData type: {daily_sales.dtype}")
print(f"Number of elements: {len(daily_sales)}")
````

````python
# Creating a Series with a custom index
# Now the data has meaningful labels!
daily_sales = pd.Series(
    [250, 180, 320, 290, 410],
    index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    name='Daily Sales ($)'
)
print("Daily Sales with Custom Index:")
print(daily_sales)
````

````python
# Creating a Series from a dictionary (keys become index)
# This is a natural way to create labeled data
inventory = pd.Series({
    'Laptop': 45,
    'Mouse': 200,
    'Keyboard': 150,
    'Monitor': 30,
    'Webcam': 75
})
print("Inventory:")
print(inventory)
````

### Accessing Series Elements

You can access elements by label (like a dictionary) or by position (like a list).

````python
# Access by label
print(f"Monday sales: ${daily_sales['Mon']}")
print(f"Friday sales: ${daily_sales['Fri']}")

# Access multiple labels
print("\nWeekend prep days (Thu, Fri):")
print(daily_sales[['Thu', 'Fri']])
````

````python
# Access by position using .iloc
print(f"First day's sales: ${daily_sales.iloc[0]}")
print(f"Last day's sales: ${daily_sales.iloc[-1]}")

# Slicing by position
print("\nFirst 3 days:")
print(daily_sales.iloc[:3])
````

### Series Operations

Like NumPy arrays, Series support vectorized operations—no loops needed!

````python
# Vectorized arithmetic
print("Original sales:")
print(daily_sales)

# Calculate 10% bonus on all sales
bonus = daily_sales * 0.10
print("\n10% Bonus:")
print(bonus)
````

````python
# Series methods for statistics
print(f"Total weekly sales: ${daily_sales.sum()}")
print(f"Average daily sales: ${daily_sales.mean():.2f}")
print(f"Best day: ${daily_sales.max()} ({daily_sales.idxmax()})")
print(f"Worst day: ${daily_sales.min()} ({daily_sales.idxmin()})")
print(f"Standard deviation: ${daily_sales.std():.2f}")
````

````python
# Boolean filtering on Series
# Find days with above-average sales
avg = daily_sales.mean()
above_average = daily_sales[daily_sales > avg]
print(f"Days with sales above ${avg:.0f}:")
print(above_average)
````

## 8.3 DataFrame: The Two-Dimensional Powerhouse

A **DataFrame** is a two-dimensional labeled data structure—think of it as a spreadsheet or SQL table in Python. It's the primary Pandas object you'll work with.

### Anatomy of a DataFrame

```
         Column1    Column2    Column3
Index                                  
  0        value      value      value
  1        value      value      value
  2        value      value      value
```

Key components:
- **Columns**: Named, can hold different data types
- **Index**: Row labels (default: 0, 1, 2, ...)
- **Data**: The actual values in the table

**Important insight**: A DataFrame is essentially a collection of Series that share the same index.

````python
# Creating a DataFrame from a dictionary
# Each key becomes a column name, each list becomes column values
products = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics'],
    'Price': [999.99, 29.99, 79.99, 299.99, 89.99],
    'Stock': [50, 200, 150, 75, 100],
    'Supplier': ['TechCorp', 'AccessoriesPlus', 'AccessoriesPlus', 'TechCorp', 'TechCorp']
})

print("Product Inventory:")
products
````

````python
# Essential DataFrame attributes
print(f"Shape (rows, columns): {products.shape}")
print(f"Number of rows: {len(products)}")
print(f"Number of columns: {len(products.columns)}")
print(f"\nColumn names: {products.columns.tolist()}")
print(f"Index: {products.index.tolist()}")
````

````python
# Data types in each column
print("Data Types:")
print(products.dtypes)
print(f"\nNote: 'object' means string/text data in Pandas")
````

### Quick Data Exploration

When you first load a dataset, these methods help you understand what you're working with:

````python
# View first few rows
print("First 3 rows:")
products.head(3)
````

````python
# View last few rows
print("Last 2 rows:")
products.tail(2)
````

## 8.5 Selecting Data: The Art of Data Access

Selecting subsets of data is one of the most important skills in data analysis. Pandas provides multiple ways to access data, each suited to different situations.

### Method Overview

| Method | Purpose | Syntax |
|--------|---------|--------|
| `[]` | Select columns | `df['column']` or `df[['col1', 'col2']]` |
| `.loc[]` | Label-based selection | `df.loc[row_labels, col_labels]` |
| `.iloc[]` | Position-based selection | `df.iloc[row_positions, col_positions]` |
| Boolean indexing | Conditional selection | `df[df['column'] > value]` |

````python
# Create sample data for demonstrations
employees = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward', 'Fiona'],
    'Department': ['Sales', 'IT', 'Sales', 'HR', 'IT', 'Marketing'],
    'Salary': [65000, 78000, 58000, 52000, 85000, 71000],
    'Years': [3, 5, 2, 4, 7, 3],
    'Rating': [4.2, 3.8, 4.5, 3.5, 4.8, 4.1]
})
print("Employee Data:")
employees
````

### 8.5.1 Selecting Columns

````python
# Select a single column (returns a Series)
names = employees['Name']
print("Names (as Series):")
print(names)
print(f"\nType: {type(names)}")
````

````python
# Select multiple columns (returns a DataFrame)
# Note: Use double brackets for multiple columns!
name_salary = employees[['Name', 'Salary']]
print("Names and Salaries (as DataFrame):")
print(name_salary)
print(f"\nType: {type(name_salary)}")
````

````python
# Dot notation (alternative for simple column names)
# Note: Only works if column name has no spaces or special characters
print("Salaries using dot notation:")
print(employees.Salary)
````

### 8.5.2 Selecting Rows with loc and iloc

**Key Distinction:**
- `loc`: Uses **labels** (what you see in the index/columns)
- `iloc`: Uses **integer positions** (0, 1, 2, ...)

````python
# loc: Label-based selection
print("Row at index label 0:")
print(employees.loc[0])
````

````python
# loc: Multiple rows by label
print("Rows 0, 2, and 4:")
employees.loc[[0, 2, 4]]
````

````python
# loc: Select specific rows AND columns
print("Rows 0-2, Name and Salary columns:")
employees.loc[0:2, ['Name', 'Salary']]
````

````python
# iloc: Position-based selection
print("First row (position 0):")
print(employees.iloc[0])
````

````python
# iloc: Slicing (exclusive of end index, like Python lists)
print("First 3 rows:")
employees.iloc[:3]
````

````python
# iloc: Select specific rows and columns by position
print("First 3 rows, first 2 columns:")
employees.iloc[:3, :2]
````

````python
# iloc: Last row (negative indexing works!)
print("Last row:")
print(employees.iloc[-1])
````

### 8.5.3 Boolean Filtering

Boolean filtering is one of the most powerful features in Pandas. It lets you select rows based on conditions—just like WHERE clauses in SQL.

````python
# Simple condition: Find employees earning > $70,000
high_earners = employees[employees['Salary'] > 70000]
print("Employees earning > $70,000:")
high_earners
````

````python
# Understanding how it works: the condition creates a boolean Series
print("Boolean mask (True where condition is met):")
print(employees['Salary'] > 70000)
````

````python
# Multiple conditions with AND (&)
# IMPORTANT: Must use parentheses around each condition!
sales_high_rating = employees[
    (employees['Department'] == 'Sales') & 
    (employees['Rating'] >= 4.0)
]
print("Sales employees with rating >= 4.0:")
sales_high_rating
````

````python
# Multiple conditions with OR (|)
it_or_senior = employees[
    (employees['Department'] == 'IT') | 
    (employees['Years'] >= 5)
]
print("IT employees OR 5+ years experience:")
it_or_senior
````

````python
# Using isin() for multiple values
tech_depts = employees[employees['Department'].isin(['IT', 'Engineering'])]
print("Tech department employees:")
tech_depts
````

````python
# Using query() method for cleaner syntax (alternative approach)
result = employees.query('Salary > 60000 and Years >= 3')
print("Using query(): Salary > $60,000 AND Years >= 3")
result
````

## 8.6 Adding and Modifying Data

DataFrames are mutable—you can add, modify, and delete rows and columns.

````python
# Create a working copy (to preserve original)
df = employees.copy()

# Add a new column with calculated values
df['Bonus'] = df['Salary'] * 0.10
print("Added Bonus column (10% of salary):")
df
````

````python
# Add a column based on conditions
# Use np.where() for if-else logic
df['Performance'] = np.where(df['Rating'] >= 4.0, 'High', 'Standard')
print("Added Performance column:")
df
````

````python
# Modify values based on condition
# Give 15% bonus to employees with 5+ years (instead of 10%)
df.loc[df['Years'] >= 5, 'Bonus'] = df['Salary'] * 0.15
print("Updated bonus for senior employees:")
df[['Name', 'Years', 'Salary', 'Bonus']]
````

````python
# Calculate total compensation
df['Total_Comp'] = df['Salary'] + df['Bonus']
print("Total Compensation:")
df[['Name', 'Salary', 'Bonus', 'Total_Comp']]
````

````python
# Rename columns
df = df.rename(columns={
    'Years': 'Experience',
    'Total_Comp': 'TotalCompensation'
})
print("Renamed columns:")
print(df.columns.tolist())
````

````python
# Drop columns
df = df.drop(columns=['Performance'])
print("After dropping 'Performance' column:")
print(df.columns.tolist())
````

````python
# Add a new row using concat (pd.append is deprecated)
new_employee = pd.DataFrame([{
    'Name': 'Grace',
    'Department': 'Finance',
    'Salary': 68000,
    'Experience': 4,
    'Rating': 4.3,
    'Bonus': 6800,
    'TotalCompensation': 74800
}])

df = pd.concat([df, new_employee], ignore_index=True)
print("Added new employee:")
df
````

## 8.7 Handling Missing Data

Real-world data often has missing values. Pandas represents missing data as `NaN` (Not a Number) or `None`. Understanding how to handle missing data is crucial for accurate analysis.

### Common Causes of Missing Data in Business:
- Customer didn't answer survey questions
- System errors during data collection
- Data not applicable (e.g., "spouse name" for single employees)
- Records merged from different sources

````python
# Create data with missing values
survey = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4, 5, 6],
    'Age': [25, np.nan, 35, 42, np.nan, 31],
    'Satisfaction': [4, 5, np.nan, 3, 4, np.nan],
    'Purchases': [150, 200, 175, np.nan, 225, 180]
})

print("Survey Data with Missing Values:")
survey
````

````python
# Detect missing values
print("Missing values per column:")
print(survey.isnull().sum())

print(f"\nTotal missing values: {survey.isnull().sum().sum()}")
print(f"Percentage missing: {survey.isnull().sum().sum() / survey.size * 100:.1f}%")
````

````python
# Find rows with any missing values
print("Rows with missing values:")
survey[survey.isnull().any(axis=1)]
````

````python
# Option 1: Drop rows with missing values
clean_drop = survey.dropna()
print("After dropping rows with NaN:")
print(clean_drop)
print(f"\nRows remaining: {len(clean_drop)} of {len(survey)}")
````

````python
# Option 2: Fill missing values with a constant
clean_fill = survey.copy()
clean_fill['Satisfaction'] = clean_fill['Satisfaction'].fillna(3)  # Fill with neutral value
print("After filling Satisfaction NaN with 3:")
clean_fill
````

````python
# Option 3: Fill with mean (common for numeric data)
clean_mean = survey.copy()
clean_mean['Age'] = clean_mean['Age'].fillna(clean_mean['Age'].mean())
clean_mean['Purchases'] = clean_mean['Purchases'].fillna(clean_mean['Purchases'].mean())
print("After filling with column means:")
clean_mean
````

**When to use each approach:**

| Approach | When to Use |
|----------|-------------|
| Drop rows | Small percentage missing, analysis requires complete data |
| Fill with constant | Known default value, categorical data |
| Fill with mean/median | Numeric data, don't want to bias analysis |
| Fill with mode | Categorical data |
| Forward/back fill | Time series data |

## 8.9 Grouping and Aggregation

The `groupby()` operation is one of Pandas' most powerful features. It follows a **split-apply-combine** pattern:

1. **Split**: Divide data into groups based on some criteria
2. **Apply**: Perform a calculation on each group
3. **Combine**: Merge the results back together

Think of it like a pivot table in Excel, but more flexible.

````python
# Create sales data for demonstration
sales = pd.DataFrame({
    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Revenue': [1200, 800, 1500, 900, 1100, 750, 1300, 850],
    'Units': [120, 80, 150, 90, 110, 75, 130, 85],
    'Cost': [600, 400, 750, 450, 550, 375, 650, 425]
})
print("Sales Data:")
sales
````

````python
# Basic groupby: Total revenue by region
print("Total Revenue by Region:")
sales.groupby('Region')['Revenue'].sum()
````

````python
# Multiple aggregations on one column
print("Revenue Statistics by Region:")
sales.groupby('Region')['Revenue'].agg(['sum', 'mean', 'min', 'max', 'count'])
````

````python
# Group by multiple columns
print("Revenue by Region and Product:")
sales.groupby(['Region', 'Product'])['Revenue'].sum()
````

````python
# Unstack to create a pivot-table-like view
print("Pivot view - Revenue by Region and Product:")
sales.groupby(['Region', 'Product'])['Revenue'].sum().unstack()
````

````python
# Different aggregations for different columns
print("Multiple metrics by Region:")
summary = sales.groupby('Region').agg({
    'Revenue': 'sum',
    'Units': 'sum',
    'Cost': 'sum'
})

# Calculate profit margin
summary['Profit'] = summary['Revenue'] - summary['Cost']
summary['Margin_%'] = (summary['Profit'] / summary['Revenue'] * 100).round(1)
summary
````

````python
# Named aggregations (cleaner syntax)
print("Using named aggregations:")
sales.groupby('Region').agg(
    Total_Revenue=('Revenue', 'sum'),
    Avg_Revenue=('Revenue', 'mean'),
    Total_Units=('Units', 'sum')
).round(2)
````

## 8.10 Combining DataFrames

Often you need to combine data from multiple sources. Pandas provides several methods:

| Method | Use Case | SQL Equivalent |
|--------|----------|----------------|
| `concat()` | Stack DataFrames vertically or horizontally | UNION |
| `merge()` | Join on keys (like database tables) | JOIN |
| `join()` | Join on index | JOIN |

### 8.10.1 Concatenation

````python
# Read monthly revenue data from repository
jan_revenue = pd.read_csv(f"{DATA_URL}/week08/jan_revenue.csv")
feb_revenue = pd.read_csv(f"{DATA_URL}/week08/feb_revenue.csv")
mar_revenue = pd.read_csv(f"{DATA_URL}/week08/mar_revenue.csv")

# Add month column to identify source
jan_revenue['Month'] = 'January'
feb_revenue['Month'] = 'February'
mar_revenue['Month'] = 'March'

print("January:")
print(jan_revenue)
print("\nFebruary:")
print(feb_revenue)
````

````python
# Concatenate vertically (stack rows)
quarterly = pd.concat([jan_revenue, feb_revenue, mar_revenue], ignore_index=True)
print("Combined Quarterly Data:")
quarterly
````

````python
# Now we can analyze across all months!
print("Revenue by Product:")
quarterly.groupby('Product')['Revenue'].agg(['sum', 'mean']).round(2)
````

### 8.10.2 Merging (Joining)

````python
# Read related data from repository
orders = pd.read_csv(f"{DATA_URL}/week08/orders.csv")
customers = pd.read_csv(f"{DATA_URL}/week08/customers.csv")

print("Orders:")
display(orders)
print("\nCustomers:")
display(customers)
````

````python
# Merge on CustomerID (inner join by default)
merged = pd.merge(orders, customers, on='CustomerID')
print("Merged Orders with Customer Info:")
merged
````

````python
# Now we can analyze orders with customer details!
print("Total Orders by Customer:")
merged.groupby('Name').agg(
    Order_Count=('OrderID', 'count'),
    Total_Amount=('Amount', 'sum'),
    Avg_Order=('Amount', 'mean')
).round(2)
````

````python
# Orders by City
print("\nOrders by City:")
merged.groupby('City')['Amount'].agg(['count', 'sum', 'mean']).round(2)
````

### Merge Types

```python
pd.merge(left, right, on='key', how='inner')  # Only matching rows (default)
pd.merge(left, right, on='key', how='left')   # All left rows, matching right
pd.merge(left, right, on='key', how='right')  # All right rows, matching left
pd.merge(left, right, on='key', how='outer')  # All rows from both
```

````python
# Overall KPIs
print("=" * 60)
print("RETAIL SALES PERFORMANCE REPORT")
print("=" * 60)

print("\n📊 KEY PERFORMANCE INDICATORS")
print("-" * 40)
print(f"Total Sales Revenue:     ${retail_sales['Sales'].sum():,}")
print(f"Average Sale per Entry:  ${retail_sales['Sales'].mean():,.2f}")
print(f"Total Transactions:      {retail_sales['Transactions'].sum():,}")
print(f"Average Ticket Size:     ${retail_sales['Sales'].sum() / retail_sales['Transactions'].sum():.2f}")
````

````python
# Store Performance Analysis
print("\n🏪 STORE PERFORMANCE")
print("-" * 40)

store_summary = retail_sales.groupby('Store').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Transactions=('Transactions', 'sum')
)
store_summary['Avg_Ticket'] = (store_summary['Total_Sales'] / 
                                store_summary['Total_Transactions']).round(2)
store_summary['Sales_Share_%'] = (store_summary['Total_Sales'] / 
                                   store_summary['Total_Sales'].sum() * 100).round(1)

display(store_summary.sort_values('Total_Sales', ascending=False))
````

````python
# Category Performance Analysis
print("\n📦 CATEGORY PERFORMANCE")
print("-" * 40)

category_summary = retail_sales.groupby('Category').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Transactions=('Transactions', 'sum')
)
category_summary['Avg_Ticket'] = (category_summary['Total_Sales'] / 
                                   category_summary['Total_Transactions']).round(2)
category_summary['Sales_Share_%'] = (category_summary['Total_Sales'] / 
                                      category_summary['Total_Sales'].sum() * 100).round(1)

display(category_summary.sort_values('Total_Sales', ascending=False))
````

````python
# Store x Category Performance Matrix
print("\n📈 SALES BY STORE AND CATEGORY")
print("-" * 40)

pivot = retail_sales.pivot_table(
    values='Sales', 
    index='Store', 
    columns='Category', 
    aggfunc='sum'
)
pivot['TOTAL'] = pivot.sum(axis=1)
display(pivot)
````

````python
# Calculate metrics per store-category combination
retail_sales['Avg_Transaction'] = retail_sales['Sales'] / retail_sales['Transactions']

# Find best performing combinations
print("\n🏆 TOP PERFORMING STORE-CATEGORY COMBINATIONS")
print("-" * 40)
print("(By Total Sales)")
display(retail_sales.nlargest(5, 'Sales')[['Store', 'Category', 'Sales', 'Transactions']])
````

````python
# Identify opportunities (lowest performers)
print("\n⚠️ AREAS NEEDING ATTENTION")
print("-" * 40)
print("(Lowest Sales by Category-Store)")
display(retail_sales.nsmallest(3, 'Sales')[['Store', 'Category', 'Sales', 'Transactions']])
````

````python
# Executive Summary
print("\n" + "=" * 60)
print("EXECUTIVE SUMMARY")
print("=" * 60)

best_store = store_summary['Total_Sales'].idxmax()
best_category = category_summary['Total_Sales'].idxmax()

print(f"\n📍 Best Performing Store: {best_store}")
print(f"   Revenue: ${store_summary.loc[best_store, 'Total_Sales']:,}")
print(f"   Market Share: {store_summary.loc[best_store, 'Sales_Share_%']}%")

print(f"\n📦 Best Performing Category: {best_category}")
print(f"   Revenue: ${category_summary.loc[best_category, 'Total_Sales']:,}")
print(f"   Share of Sales: {category_summary.loc[best_category, 'Sales_Share_%']}%")

highest_ticket = category_summary['Avg_Ticket'].idxmax()
print(f"\n💰 Highest Ticket Category: {highest_ticket}")
print(f"   Average Transaction: ${category_summary.loc[highest_ticket, 'Avg_Ticket']}")
````

````python
# Salary Analysis
print("\n💰 SALARY ANALYSIS")
print("-" * 40)
print(f"Total Payroll: ${employees['Salary'].sum():,}")
print(f"Average Salary: ${employees['Salary'].mean():,.2f}")
print(f"Median Salary: ${employees['Salary'].median():,.2f}")
print(f"Salary Range: ${employees['Salary'].min():,} - ${employees['Salary'].max():,}")
````

````python
# Department Breakdown
print("\n🏢 DEPARTMENT BREAKDOWN")
print("-" * 40)

dept_summary = employees.groupby('Department').agg(
    Employees=('Name', 'count'),
    Avg_Salary=('Salary', 'mean'),
    Avg_Years=('Years', 'mean'),
    Avg_Rating=('Rating', 'mean')
).round(2)

dept_summary['Salary_Budget'] = employees.groupby('Department')['Salary'].sum()
display(dept_summary)
````

````python
# Performance Tiers
print("\n⭐ PERFORMANCE DISTRIBUTION")
print("-" * 40)

# Categorize employees by rating
def rating_tier(rating):
    if rating >= 4.5:
        return 'Exceptional'
    elif rating >= 4.0:
        return 'Strong'
    elif rating >= 3.5:
        return 'Solid'
    else:
        return 'Developing'

employees['Performance_Tier'] = employees['Rating'].apply(rating_tier)
tier_counts = employees['Performance_Tier'].value_counts()

for tier, count in tier_counts.items():
    pct = count / len(employees) * 100
    print(f"{tier:12}: {count} employees ({pct:.0f}%)")
````

````python
# Tenure vs Salary Analysis
print("\n📊 TENURE-SALARY ANALYSIS")
print("-" * 40)

# Bin employees by tenure
employees['Tenure_Band'] = pd.cut(
    employees['Years'],
    bins=[0, 2, 5, float('inf')],
    labels=['< 2 years', '2-5 years', '5+ years']
)

tenure_analysis = employees.groupby('Tenure_Band').agg(
    Count=('Name', 'count'),
    Avg_Salary=('Salary', 'mean'),
    Avg_Rating=('Rating', 'mean')
).round(2)

display(tenure_analysis)
````

````python
# Top Performers
print("\n🏆 TOP PERFORMERS")
print("-" * 40)
top_performers = employees.nlargest(3, 'Rating')[['Name', 'Department', 'Rating', 'Years', 'Salary']]
display(top_performers)
````

## 8.14 Best Practices for Pandas

As you develop your Pandas skills, keep these best practices in mind:

### 1. Always Explore Your Data First
```python
df.head()      # First rows
df.info()      # Data types and nulls
df.describe()  # Statistics
df.shape       # Dimensions
```

### 2. Use Meaningful Variable Names
```python
# Good
employee_salaries = df['Salary']
high_performers = df[df['Rating'] > 4.5]

# Avoid
x = df['Salary']
filtered = df[df['Rating'] > 4.5]
```

### 3. Chain Operations When Appropriate
```python
# Chain for clarity
result = (df
    .query('Salary > 50000')
    .groupby('Department')['Rating']
    .mean()
    .round(2))
```

### 4. Use .copy() When Modifying Subsets
```python
# Avoid SettingWithCopyWarning
subset = df[df['Year'] == 2024].copy()
subset['NewColumn'] = 'value'
```

### 5. Handle Missing Data Intentionally
```python
# Check for nulls first
print(df.isnull().sum())

# Then decide on strategy
df['Column'].fillna(df['Column'].mean(), inplace=True)
```

## 8.15 Data Quality and Governance: The Foundation of Trustworthy Analytics

As you develop your data analysis skills with Pandas, it's essential to understand that **technical proficiency is only part of the story**. In real-world business settings, you'll encounter questions that go beyond "Can I manipulate this data?" to "Should I? How should I document my decisions? Who is responsible for data quality?"

### Why Data Governance Matters

Consider these scenarios:
- A marketing analyst discovers customer email addresses in a dataset. Can they use this for a new campaign?
- A sales report shows revenue numbers that seem too high. How do you verify the data is accurate?
- An AI tool generates a customer churn prediction. How was the data collected? Are there biases?

These questions touch on **data governance**—the framework of policies, procedures, and practices that ensure data is accurate, secure, and used appropriately.

### The Five Dimensions of Data Quality

Before you analyze data, assess its quality across these dimensions:

| Dimension | Question to Ask | Example Problem |
|-----------|-----------------|-----------------|
| **Accuracy** | Is the data factually correct? | Customer age listed as 150 years |
| **Completeness** | Is all necessary data present? | 30% of records missing phone numbers |
| **Consistency** | Is data uniform across sources? | "NY" vs "New York" vs "N.Y." in address field |
| **Timeliness** | Is the data current enough? | Using 2020 pricing data for 2026 decisions |
| **Validity** | Does data conform to rules? | Email addresses without @ symbol |

### Data Quality Assessment in Practice

```python
# Always document your data quality assessment
def assess_data_quality(df):
    """Generate a data quality report for a DataFrame."""
    report = {
        'total_rows': len(df),
        'total_columns': len(df.columns),
        'missing_values': df.isnull().sum().sum(),
        'missing_percentage': (df.isnull().sum().sum() / df.size * 100),
        'duplicate_rows': df.duplicated().sum()
    }
    return report

# Document any data quality issues you find
# This creates an audit trail for your analysis
```

### The Data Lineage Question

**Data lineage** answers: "Where did this data come from, and what happened to it?"

In your work, you should always be able to explain:
1. **Source**: Where did the original data come from?
2. **Transformations**: What cleaning, filtering, or calculations were applied?
3. **Assumptions**: What decisions did you make and why?

```python
# Good practice: Document transformations
# Original data source: Sales database, extracted 2026-01-15
# Transformation 1: Removed records with negative sales (3 rows, assumed errors)
# Transformation 2: Filled missing regions based on store ID mapping
# Transformation 3: Converted dates from string to datetime format
```

### Privacy and Ethical Considerations

When working with data, especially in business contexts, be aware of:

**Personally Identifiable Information (PII):**
- Names, addresses, phone numbers, email addresses
- Social Security numbers, financial account numbers
- Any data that could identify an individual

**Best Practices:**
- Never include PII in code comments or variable names
- Be cautious when sharing notebooks or datasets
- Ask: "Do I need this data for my analysis?"
- Consider anonymization or aggregation when possible

**Sensitive Business Data:**
- Salary information
- Customer lists
- Strategic plans
- Pricing models

### The AI Connection: Verifying AI-Assisted Data Work

As AI tools become more prevalent in data analysis, governance becomes even more important:

| AI Assistance | Your Responsibility |
|---------------|---------------------|
| AI generates a cleaning script | Verify each transformation makes sense |
| AI suggests handling missing values | Understand why that approach is appropriate |
| AI creates visualizations | Confirm the representation is accurate and not misleading |
| AI summarizes data patterns | Validate findings against the actual data |

**Key Insight:** AI can process data faster, but **you** are responsible for the accuracy, appropriateness, and ethical use of the results.

### Documentation as Professional Practice

Strong analysts document their work. For each analysis, record:

1. **Data Source Documentation**
   - Where data came from
   - When it was extracted
   - Who provided access

2. **Data Quality Notes**
   - Issues found
   - Decisions made
   - Assumptions applied

3. **Transformation Log**
   - What changes were made
   - Why each change was necessary
   - What was excluded and why

4. **Verification Steps**
   - How accuracy was checked
   - What sanity checks were performed
   - Any discrepancies found

### A Framework for Responsible Data Analysis

Before beginning any analysis, ask yourself:

1. **Authorization**: Am I authorized to access and use this data?
2. **Purpose**: Is my use aligned with the data's intended purpose?
3. **Quality**: Have I assessed and documented data quality issues?
4. **Privacy**: Am I handling personal information appropriately?
5. **Transparency**: Can I explain my process to a stakeholder?
6. **Verification**: How will I validate my results?

### Business Impact of Poor Data Governance

| Issue | Business Impact |
|-------|-----------------|
| Using inaccurate data | Wrong business decisions, lost revenue |
| Privacy violations | Legal penalties, reputation damage |
| Undocumented transformations | Inability to reproduce or explain results |
| Biased datasets | Discriminatory outcomes, ethical failures |
| No quality checks | Reports that stakeholders can't trust |

### Looking Forward

As you progress through this course and into your career, you'll find that the analysts who advance fastest are those who combine technical skill with strong governance practices. Anyone can run a `groupby()` operation—the professionals are those who can explain their data sources, justify their decisions, and ensure their work is accurate, ethical, and reproducible.

The habits you develop now—documenting your work, verifying data quality, respecting privacy—will distinguish you as a trustworthy analyst in any organization.
