## Exercise 2: Revenue Calculation (2 points)

A company sells software subscriptions. Calculate the monthly recurring revenue (MRR) based on:

- Number of Basic subscribers: 150 at $29.99/month
- Number of Pro subscribers: 75 at $79.99/month
- Number of Enterprise subscribers: 20 at $199.99/month

Calculate and print:
1. Revenue from each tier
2. Total MRR
3. Projected Annual Recurring Revenue (ARR = MRR * 12)

````python
# Exercise 2: 

# Define subscriber counts and prices
basic_subscribers = 150
basic_price = 29.99

pro_subscribers = 75
pro_price = 79.99

enterprise_subscribers = 20
enterprise_price = 199.99



# Calculated revenue per tier
basic_plan_rev = basic_subscribers * basic_price 

pro_plan_rev = pro_subscribers * pro_price

enterprise_plan_rev = enterprise_subscribers * enterprise_price

# Calculate total MRR
MRR = basic_plan_rev + pro_plan_rev + enterprise_plan_rev 

# Calculate ARR
ARR = MRR * 12

# Print results
print(f"Basic Plan Revenue = ${basic_plan_rev}")
print()
print(f"Pro Plan Revenue = ${pro_plan_rev}")
print()
print(f"Enterprise Plan Revenue = ${enterprise_plan_rev}")
print()
print(f"Monthly Recurring Revenue = ${MRR}")
print()
print(f"Annual Recurring Revenue = ${ARR}")

````

## Exercise 4: Data Type Conversion (2 points)

You receive the following data as strings from a web form:

```python
product_price_str = "149.99"
quantity_str = "3"
tax_rate_str = "0.07"
```

Convert these strings to appropriate numeric types and calculate:
1. Subtotal (price * quantity)
2. Tax amount (subtotal * tax_rate)
3. Total (subtotal + tax)

Print the final total formatted as a string: "Your total is: $XXX.XX"

````python
# Exercise 4: Your code here

# Given string data
product_price_str = "149.99"
quantity_str = "3"
tax_rate_str = "0.07"

# Convert to numeric types
product_price_float = float(product_price_str)
quantity_int = int(quantity_str)
tax_rate_float = float(tax_rate_str)

# Calculate subtotal, tax, and total
subtotal = product_price_float * quantity_int

tax = subtotal * tax_rate_float

Total = subtotal + tax

# Print formatted result
print(f"Your total is: {Total}")
````

## Exercise 5: Business Metrics Dashboard (2 points)

Create a simple text-based dashboard that displays key business metrics. Use variables, arithmetic operators, and comparison operators.

Given data:
- Current month sales: $125,000
- Previous month sales: $110,000
- Sales target: $120,000
- Number of transactions: 450

Calculate and display:
1. Month-over-month growth rate (as a percentage)
2. Whether the sales target was met (True/False)
3. Average transaction value
4. Amount above or below target

Format your output nicely using print statements with descriptive labels.

````python
# Exercise 5: 

# Define the given data
CurrentMonth_sales = 125000
PreviousMonth_sales = 110000
target_sales = 120000
transaction_n = 450

# Calculate metrics

# 1. Month-over-month growth rate (as a percentage)
MOM_growth = (CurrentMonth_sales - PreviousMonth_sales) / PreviousMonth_sales * 100

# 2. Was the sales target met?
currentMonthSalesTarget = CurrentMonth_sales > target_sales

previousMonthSalesTarget = PreviousMonth_sales > target_sales

# 3. Average Transaction Value (I am assuming that the transactions value given is an accumulated sum of transactions from previous and current month)
avgTransactionValue = (CurrentMonth_sales + PreviousMonth_sales) // transaction_n

#4 Amount above or below target ()
prev_month_difference_from_target = PreviousMonth_sales - target_sales

current_month_difference_from_target = CurrentMonth_sales - target_sales


# Print dashboard
print("="*40)
print("     MONTHLY SALES DASHBOARD")
print("="*40)

print(f"Month-over-Month Growth: {MOM_growth}%")
print(""" """)

print(f"Current month sales number met sales target: {currentMonthSalesTarget}")
print(f"Previous month sales number met sales target: {previousMonthSalesTarget}")

print(""" """)

print(f"Average transaction value: ${avgTransactionValue}")

print(""" """)


print(f"Amount above/below sales target for previous month: ${prev_month_difference_from_target}")
print(f"Amount above/below sales target for current month: ${current_month_difference_from_target}")
````
