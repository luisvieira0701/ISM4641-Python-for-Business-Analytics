## Part 2: Revenue Analysis (25 points)

David pulls up NovaTech's subscription data. *"This is the heart of our business model,"* he explains. *"We have three pricing tiers, and investors will want to understand our revenue composition."*

**Subscription Tiers:**

| Tier | Monthly Price | Number of Customers |
|------|--------------|--------------------|
| Starter | $49.00 | 312 |
| Professional | $149.00 | 156 |
| Enterprise | $499.00 | 38 |

**Your Task:**
1. Calculate Monthly Recurring Revenue (MRR) for each tier
2. Calculate Total MRR
3. Calculate Annual Recurring Revenue (ARR) = MRR Ã— 12
4. Calculate the percentage of total MRR that comes from each tier
5. Calculate the total number of customers
6. Calculate Average Revenue Per User (ARPU) = Total MRR / Total Customers

**Display a formatted revenue report that Marcus can use in the pitch.**

```python
# Part 2: Revenue Analysis

# Calculate NovaTech's key revenue metrics

# Subscription tier data
Starter_Price = 49
Starter_Customers = 312
Professional_Price = 149
Professional_Customers = 156
Enterprise_Price = 499
Enterprise_Customers = 38


# Calculate MRR per tier
Starter_MRR = Starter_Price * Starter_Customers # Integers multiplication to calculate MRR (starter)
Professional_MRR = Professional_Price * Professional_Customers # Integers multiplication to calculate MRR (professional)
Enterprise_MRR = Enterprise_Price * Enterprise_Customers # Integers multiplication to calculate MRR (enterprise)

# Calculate total MRR and ARR
Total_MRR = Starter_MRR + Professional_MRR + Enterprise_MRR
Total_ARR = Total_MRR * 12 # As refered in question, ARR is calculated by multiplying MRR by 12


# Calculate percentages
Starter_percentage = (Starter_MRR / Total_MRR) * 100
Professional_percentage = (Professional_MRR / Total_MRR) * 100
Enterprise_percentage = (Enterprise_MRR / Total_MRR) * 100

# Calculate customer metrics
Total_Customers = Starter_Customers + Professional_Customers + Enterprise_Customers
ARPU = Total_MRR / Total_Customers

# Print formatted revenue report
print("="*60)
print("              REVENUE ANALYSIS") ## Header for revenue alaysis
print()
print(f"Key Metrics:") ## Section for overall metrics formatted with dashes for clear visual
print()

print(f"Total MRR: ${Total_MRR:,.2f}")  ##Using commas as thousand separators
print(f"Total ARR: ${Total_ARR:,.2f}")
print(f"Total Customers: {Total_Customers:,}")
print(f"ARPU: ${ARPU:.2f}")  ##Using f-strings to format output
print("-"*60)


print("Starter Tier Information")
print()

print(f"Number of Starter Tier Customers: {Starter_Customers}")
print(f"Starter Tier MRR: ${Starter_MRR:,.2f}")   ##Separating metrics per tier
print(f"Percentage of Total MRR from Starter Tier: {Starter_percentage:.2f}%")


print("-"*60) ##Using dashes to separate sections

print("Professional Tier Information")
print()
print(f"Number of Professional Tier Customers: {Professional_Customers}")
print(f"Professional Tier MRR: ${Professional_MRR:,.2f}")
print(f"Percentage of Total MRR from Professional Tier: {Professional_percentage:.2f}%")
print("-"*60)

print("Enterprise Tier Information")
print()
print(f"Number of Enterprise Tier Customers: {Enterprise_Customers}")
print(f"Enterprise Tier MRR: ${Enterprise_MRR:,.2f}")
print(f"Percentage of Total MRR from Enterprise Tier: {Enterprise_percentage:.2f}%")
print("-"*60)

print("="*60)
# Add your formatted output here
```

```python
# Part 3: Growth Metrics
# Analyze NovaTech's growth trajectory

# Monthly data  
#October 2025 Data
October_New_Customers = 28
October_Churned_Customers = 5
October_MRR_MonthEnd = 68420

# November 2025 Data
November_New_Customers = 35
November_Churned_Customers = 7
November_MRR_MonthEnd = 71850                   ## Displaying data for all months

#December 2025 Data
December_New_Customers = 42
December_Churned_Customers = 4
December_MRR_MonthEnd = 76290

# January 2026 Data
January_New_Customers = 51
January_Churned_Customers = 6
January_MRR_MonthEnd = 81634


# 1. Calculate net new customers per month
# October net new customers
October_Net_New_Customers = October_New_Customers - October_Churned_Customers

# November net new customers
November_Net_New_Customers = November_New_Customers - November_Churned_Customers            # Basic arithmetic operation for monthly net new customers

# December net new customers
December_Net_New_Customers = December_New_Customers - December_Churned_Customers

# January net new customers
January_Net_New_Customers = January_New_Customers - January_Churned_Customers


# 2. Calculate MRR growth rate
MRR_Growth_Rate = ((January_MRR_MonthEnd - October_MRR_MonthEnd) / October_MRR_MonthEnd) * 100  # MRR growth rate calculation using basic arithmetic operations

# 3. Total Customers acquired over last 4 months
Total_New_Customers_Acquired = (October_Net_New_Customers + November_Net_New_Customers + December_Net_New_Customers + January_Net_New_Customers)    ## 4 last month customers acquired information

# 4. Total Customers Churned over last 4 months
Total_Customers_Churned = (October_Churned_Customers + November_Churned_Customers + December_Churned_Customers + January_Churned_Customers)         ##Same as above

# 5. Average monthly churn rate
Average_Monthly_Churn_Rate = (Total_Customers_Churned / Total_New_Customers_Acquired) * 100

# 6. Determine if "growing fast" 
Is_NovaTech_Growing = (Total_New_Customers_Acquired / 4) > 20 ## Conditional Boolean stroing growth status (better than conditional statements for reusability)

    


# Print growth summary
print("="*60)
print("              GROWTH METRICS")
print("         October 2025 - January 2026")
print(

)
if Is_NovaTech_Growing == True: ## Conditional statement to determine growth status for better readability
    print(f"NovaTech is growing fast with an average of {Total_New_Customers_Acquired / 4:.2f} new customers per month.")
else:
     print(f"NovaTech is not growing fast, averaging {Total_New_Customers_Acquired / 4:.2f} new customers per month.")
print(

)
print(f"Net New Customers per Month:")
print(
    ##Separating print for better visibility
)
print(f"October: {October_Net_New_Customers}")
print(f"November: {November_Net_New_Customers}")
print(f"December: {December_Net_New_Customers}")
print(f"January: {January_Net_New_Customers}")
print(
)
print(f"Total Net New Customers Acquired (Oct-Jan): {Total_New_Customers_Acquired}")
print(f"Total Customers Churned (Oct-Jan): {Total_Customers_Churned}")
print(

)
print(f"MRR Growth Rate (Oct-Jan): {MRR_Growth_Rate:.2f}%")
print(f"Average Monthly Churn Rate (Oct-Jan): {Average_Monthly_Churn_Rate:.2f}%") ## Using :.2f to format percentage outputs

print("="*60)
# Add your formatted output here
```
