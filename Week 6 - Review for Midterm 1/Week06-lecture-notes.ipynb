### Variables, Data Types & F-Strings (Week 1)

Every analysis starts with data stored in variables. Python has four basic data types you use constantly:

| Type | Example | Use Case |
|------|---------|----------|
| `str` | `"Laptop"` | Names, labels, text |
| `int` | `150` | Counts, indices, whole numbers |
| `float` | `29.99` | Prices, percentages, measurements |
| `bool` | `True` / `False` | Flags, conditions, filters |

Key skills: creating variables, converting between types (`int()`, `float()`, `str()`, `bool()`), arithmetic operators (`+`, `-`, `*`, `/`, `//`, `%`, `**`), and formatting output with f-strings.

```python
# Week 1 Recap: Variables, types, and f-strings
product = "Wireless Mouse"       # str
price = 29.99                    # float
quantity = 150                   # int
in_stock = True                  # bool

# f-strings for formatted output
print(f"{product}: ${price:.2f} x {quantity} = ${price * quantity:,.2f}")
# Wireless Mouse: $29.99 x 150 = $4,498.50

# Type checking
print(f"\nType of price: {type(price).__name__}")     # float
print(f"Type of quantity: {type(quantity).__name__}")  # int
print(f"Type of in_stock: {type(in_stock).__name__}")  # bool
```

```python
# Type conversion — essential for cleaning data
user_input = "42"
print(f"String to int: int('42') = {int(user_input)}")
print(f"Int to float:  float(42) = {float(42)}")
print(f"Float to int:  int(7.9)  = {int(7.9)}")    # Truncates, does NOT round
print(f"Int to string: str(100)  = '{str(100)}'")
print(f"Bool to int:   int(True) = {int(True)}")    # True=1, False=0
print(f"Bool of zero:  bool(0)   = {bool(0)}")      # 0 is False, everything else is True

# Arithmetic operators
print(f"\n10 / 3  = {10 / 3}")        # 3.333... (true division)
print(f"10 // 3 = {10 // 3}")         # 3 (floor division — integer result)
print(f"10 % 3  = {10 % 3}")          # 1 (modulus — remainder)
print(f"2 ** 5  = {2 ** 5}")          # 32 (exponentiation)
```

```python
# F-string formatting options you should know
amount = 1234567.891
pct = 0.1547

print(f"Default:           {amount}")
print(f"2 decimal places:  {amount:.2f}")         # 1234567.89
print(f"Commas + decimals: {amount:,.2f}")         # 1,234,567.89
print(f"Dollar format:     ${amount:,.2f}")        # $1,234,567.89
print(f"Percentage:        {pct:.1%}")             # 15.5%
print(f"Right-aligned:     '{amount:>15,.2f}'")    # right-justify in 15 chars
print(f"Left-aligned:      '{amount:<15,.2f}'")
```

### Control Flow (Week 3)

Control flow lets you make decisions (`if`/`elif`/`else`) and repeat actions (`for`, `while`). These are essential for categorizing data and processing collections.

Key patterns to remember:
- **`if`/`elif`/`else`** — choose one path based on conditions (only ONE block executes)
- **`for` loops** — iterate over a known collection (`range()`, lists, dicts)
- **`while` loops** — repeat until a condition becomes `False`
- **`break`** — exit a loop immediately
- **`continue`** — skip the rest of this iteration, move to the next
- **`range(start, stop, step)`** — generates a sequence of numbers (stop is excluded)

```python
# Week 3 Recap: Control flow

# if/elif/else — classify data based on conditions
revenue = 55000
if revenue > 50000:
    tier = "Gold"
elif revenue > 20000:
    tier = "Silver"
else:
    tier = "Bronze"
print(f"Revenue ${revenue:,} -> Tier: {tier}")

# for loop — process a collection of dictionaries
catalog = [
    {"name": "Laptop", "price": 999},
    {"name": "Mouse", "price": 29},
    {"name": "Monitor", "price": 350}
]

for product in catalog:
    if product["price"] > 100:
        print(f"{product['name']} is premium (${product['price']})")

# range() with different arguments
print(f"\nrange(5):        {list(range(5))}")        # [0, 1, 2, 3, 4]
print(f"range(2, 8):     {list(range(2, 8))}")      # [2, 3, 4, 5, 6, 7]
print(f"range(0, 10, 3): {list(range(0, 10, 3))}")  # [0, 3, 6, 9]
```

```python
# while loop — repeat until a condition is met
balance = 1000
months = 0
while balance < 2000:
    balance *= 1.05   # 5% monthly growth
    months += 1
print(f"Took {months} months to double (final: ${balance:,.2f})")

# break — exit a loop early when you find what you need
scores = [72, 85, 91, 68, 95]
for score in scores:
    if score > 90:
        print(f"First score above 90: {score}")
        break

# continue — skip items that don't meet criteria
print("\nPassing scores only:")
for score in scores:
    if score < 75:
        continue   # Skip this iteration
    print(f"  {score}")
```

### Data Structures (Week 3)

Lists, tuples, and dictionaries are the containers you use to organize data before analysis.

| Structure | Syntax | Mutable? | Use Case |
|-----------|--------|----------|----------|
| **List** | `[1, 2, 3]` | Yes | Ordered collections you might modify |
| **Tuple** | `(1, 2, 3)` | No | Fixed data (coordinates, RGB, return values) |
| **Dictionary** | `{"key": value}` | Yes | Key-value lookups, structured records |

Key operations: list slicing (`[start:stop]`), `.append()`, `.pop()`, dict `.get()`, `.keys()`, `.values()`, `.items()`, tuple unpacking.

```python
# Week 3 Recap: Data structures

# Lists — ordered, mutable collections
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]
products.append("Webcam")       # Add to the end
print(f"Products: {products}")
print(f"First: {products[0]}")  # Index 0
print(f"Last:  {products[-1]}") # Negative index = from the end

# List slicing — [start:stop] (stop is excluded)
print(f"Slice [1:3]: {products[1:3]}")   # ['Mouse', 'Keyboard']
print(f"First 3:     {products[:3]}")     # ['Laptop', 'Mouse', 'Keyboard']
print(f"Last 2:      {products[-2:]}")    # ['Monitor', 'Webcam']
```

```python
# Dictionaries — key-value pairs for structured data
employee = {
    "name": "Sarah Chen",
    "department": "Analytics",
    "salary": 72000
}
print(f"{employee['name']} works in {employee['department']}")

# .get() is safer than [] — returns None (or a default) if key is missing
print(f"Phone: {employee.get('phone', 'Not on file')}")

# Iterating over dicts
print("\nAll fields:")
for key, value in employee.items():
    print(f"  {key}: {value}")

# Adding and modifying entries
employee["title"] = "Senior Analyst"    # Add new key
employee["salary"] = 78000              # Update existing key

# Tuples — immutable, often used for unpacking
coordinates = (27.9506, -82.4572)  # USF Tampa
lat, lon = coordinates             # Tuple unpacking
print(f"\nUSF location: lat={lat}, lon={lon}")

# Functions often return tuples for multiple values
def min_max(values):
    return min(values), max(values)

lo, hi = min_max([15, 8, 23, 4, 19])
print(f"Range: {lo} to {hi}")
```

### Functions (Week 4)

Functions are reusable blocks of logic that keep your code organized and DRY (Don't Repeat Yourself). You define them with `def`, pass data in through parameters, and get results back with `return`.

Key concepts:
- **Parameters** — inputs the function needs (required and optional/default)
- **Return values** — output the function sends back to the caller
- **Scope** — variables inside a function are *local* (not visible outside)
- **Docstrings** — document what your function does
- **Lambda** — anonymous one-line functions for quick operations
- **`map()`** — apply a function to every item in a sequence
- **`filter()`** — keep only items that pass a condition

```python
# Week 4 Recap: Functions

# Basic function with docstring
def calculate_margin(revenue, cost):
    """Calculate profit margin as a percentage."""
    profit = revenue - cost
    return (profit / revenue) * 100

margin = calculate_margin(5000, 3500)
print(f"Margin: {margin:.1f}%")  # Margin: 30.0%

# Default parameters — provide a fallback value
def apply_tax(price, rate=0.07):
    """Apply sales tax. Default rate is 7%."""
    return price * (1 + rate)

print(f"Default tax: ${apply_tax(100):.2f}")       # $107.00
print(f"Custom tax:  ${apply_tax(100, 0.085):.2f}") # $108.50

# Scope — local variables stay inside the function
def demo_scope():
    local_var = "I only exist inside this function"
    return local_var

result = demo_scope()
print(f"\nReturned: {result}")
# print(local_var)  # This would cause a NameError!
```

```python
# Lambda — anonymous one-liner functions
apply_discount = lambda price, pct: price * (1 - pct / 100)
print(f"$100 with 15% off: ${apply_discount(100, 15):.2f}")  # $85.00

# map() — apply a function to every item
prices = [29.99, 49.99, 79.99]
with_tax = list(map(lambda p: round(p * 1.07, 2), prices))
print(f"Prices with 7% tax: {with_tax}")  # [32.09, 53.49, 85.59]

# filter() — keep items that pass a test
expensive = list(filter(lambda p: p > 40, prices))
print(f"Prices above $40: {expensive}")  # [49.99, 79.99]

# *args — accept any number of positional arguments (collected as a tuple)
def total_sales(*amounts):
    """Sum any number of sales amounts."""
    return sum(amounts)

print(f"\nTotal: ${total_sales(100, 250, 75):,.2f}")       # $425.00
print(f"Total: ${total_sales(100, 250, 75, 300, 50):,.2f}")  # $775.00

# **kwargs — accept any number of keyword arguments (collected as a dict)
def build_profile(**info):
    """Build a profile from keyword arguments."""
    for key, value in info.items():
        print(f"  {key}: {value}")

print("\nEmployee profile:")
build_profile(name="Alex", role="Analyst", team="Finance")
```

### NumPy (Week 5)

NumPy gives you fast numerical operations on entire datasets without writing loops. This is the most concept-dense week, so make sure you are comfortable with:

- **Creating arrays** — `np.array()`, `np.zeros()`, `np.ones()`, `np.arange()`, `np.linspace()`
- **Array attributes** — `.shape`, `.size`, `.ndim`, `.dtype`
- **Indexing & slicing** — `arr[0]`, `arr[1:4]`, `arr[:, 1]` (2D column)
- **Vectorized operations** — `arr * 2`, `arr1 + arr2` (no loops needed)
- **Statistical functions** — `np.mean()`, `np.sum()`, `np.std()`, `np.min()`, `np.max()`, `np.median()`
- **Boolean filtering** — `arr[arr > 100]`, combine with `&` and `|`
- **Useful functions** — `np.argmax()`, `np.argmin()`, `np.argsort()`, `np.diff()`
- **2D arrays** — `axis=0` (operate down columns), `axis=1` (operate across rows), `.reshape()`, `.T`

```python
# Week 5 Recap: NumPy basics

# Creating arrays
sales = np.array([1200, 1500, 1100, 1800, 1600])
zeros = np.zeros(5)                   # [0. 0. 0. 0. 0.]
ones = np.ones((2, 3))                # 2x3 matrix of 1.0s
sequence = np.arange(0, 10, 2)        # [0, 2, 4, 6, 8]
evenly = np.linspace(0, 100, 5)       # [0, 25, 50, 75, 100]

print(f"sales:    {sales}")
print(f"zeros:    {zeros}")
print(f"arange:   {sequence}")
print(f"linspace: {evenly}")

# Array attributes
print(f"\nsales.shape: {sales.shape}")   # (5,) — 1D, 5 elements
print(f"sales.size:  {sales.size}")      # 5
print(f"sales.ndim:  {sales.ndim}")      # 1
print(f"sales.dtype: {sales.dtype}")     # int64

# Statistical functions
print(f"\nTotal:   ${np.sum(sales):,}")           # $7,200
print(f"Average: ${np.mean(sales):,.0f}")          # $1,440
print(f"Std Dev: ${np.std(sales):,.0f}")           # $245
print(f"Median:  ${np.median(sales):,.0f}")        # $1,500
print(f"Min:     ${np.min(sales):,}")              # $1,100
print(f"Max:     ${np.max(sales):,}")              # $1,800
print(f"Best month index: {np.argmax(sales)}")     # 3
```

```python
# Vectorized operations — operate on entire arrays at once
sales = np.array([1200, 1500, 1100, 1800, 1600])
discounted = sales * 0.85  # 15% discount on every element
print(f"Original:   {sales}")
print(f"Discounted: {discounted}")

# Boolean filtering — no loops needed
strong_months = sales[sales > 1400]
print(f"\nSales above $1400: {strong_months}")  # [1500 1800 1600]

# Combining conditions — use & (and) and | (or) with parentheses!
mid_range = sales[(sales >= 1200) & (sales <= 1600)]
print(f"Between $1200-$1600: {mid_range}")  # [1200 1500 1600]

# Boolean arrays count True values with sum()
print(f"Months above $1400: {np.sum(sales > 1400)}")  # 3
```

```python
# 2D arrays — rows and columns
data = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
print(f"Shape: {data.shape}")    # (2, 3) — 2 rows, 3 columns
print(f"Dimensions: {data.ndim}")  # 2

# Indexing 2D arrays
print(f"\nRow 0:      {data[0]}")       # [10 20 30]
print(f"Column 1:   {data[:, 1]}")      # [20 50]
print(f"Element:    {data[1, 2]}")      # 60

# axis parameter — the key to 2D operations
print(f"\nSum all:         {data.sum()}")           # 210
print(f"Sum axis=0 (columns): {data.sum(axis=0)}")  # [50 70 90]
print(f"Sum axis=1 (rows):    {data.sum(axis=1)}")  # [ 60 150]

# Reshape — change dimensions without changing data
arr = np.arange(1, 13)  # [1, 2, 3, ..., 12]
matrix = arr.reshape(3, 4)
print(f"\nReshaped to 3x4:\n{matrix}")

# Transpose — swap rows and columns
print(f"\nTransposed (4x3):\n{matrix.T}")
```

## 6.2 Business Scenario 1: Sales Performance Analysis

**The situation:** You are a business analyst at a retail company. Your manager has given you quarterly sales data for 4 products across 3 regions and wants a complete performance summary. This is a typical task where you need to combine data organization, functions, NumPy operations, and formatted output.

### Setting Up the Data

We start by organizing the data into NumPy arrays and Python lists. The sales data is naturally a 2D structure: rows represent regions, columns represent products.

```python
import numpy as np

products = ["Laptop", "Mouse", "Keyboard", "Monitor"]
regions = ["North", "South", "West"]

# Sales data: rows = regions, columns = products
sales = np.array([
    [45000, 12000, 8500, 32000],   # North
    [38000, 15000, 9200, 28000],   # South
    [42000, 11000, 7800, 35000]    # West
])

print(f"Sales data shape: {sales.shape}")  # (3, 4) = 3 regions, 4 products
print(f"Total data points: {sales.size}")
print()
print("Sales Data (rows=regions, columns=products):")
print(sales)
```

### Writing Analysis Functions

Your manager will ask these same kinds of questions every quarter, so write reusable functions rather than one-off calculations. Notice how each function integrates multiple concepts: it takes data as parameters, uses NumPy for calculations, and returns or prints results with f-strings.

```python
def regional_summary(sales, regions):
    """Summarize total sales by region."""
    totals = sales.sum(axis=1)  # Sum across columns (products)
    for region, total in zip(regions, totals):
        print(f"  {region}: ${total:>10,.2f}")
    return totals

def best_performer(sales, labels, axis):
    """Find the best performer along an axis."""
    totals = sales.sum(axis=axis)
    best_idx = np.argmax(totals)
    return labels[best_idx], totals[best_idx]
```

### Generating the Report

Now we put the functions to work. The key insight is that `axis=1` sums across columns (giving totals per region) and `axis=0` sums across rows (giving totals per product).

```python
print("Regional Sales Summary:")
totals = regional_summary(sales, regions)
# North: $ 97,500.00
# South: $ 90,200.00
# West:  $ 95,800.00

best_region, amount = best_performer(sales, regions, axis=1)
print(f"\nBest region: {best_region} (${amount:,.0f})")
# Best region: North ($97,500)

best_product, amount = best_performer(sales, products, axis=0)
print(f"Best product: {best_product} (${amount:,.0f})")
# Best product: Laptop ($125,000)
```

### Deeper Insights with Boolean Indexing

Boolean indexing combined with control flow lets you ask more nuanced questions about the data. Here we combine NumPy boolean operations with `if`/`elif`/`else` logic and list comprehensions.

```python
# Which product-region combos exceeded $30,000?
above_target = sales > 30000
print(f"Combos above $30,000: {np.sum(above_target)}")
# 5 combos met the target

# Classify each region's performance using control flow
region_totals = sales.sum(axis=1)
print("\nRegion Performance Classification:")
for region, total in zip(regions, region_totals):
    if total > 95000:
        status = "Exceeds target"
    elif total > 85000:
        status = "Meets target"
    else:
        status = "Below target"
    print(f"  {region}: {status} (${total:,.0f})")
# North: Exceeds target ($97,500)
# South: Meets target ($90,200)
# West: Exceeds target ($95,800)
```

## 6.3 Business Scenario 2: Financial Metrics

**The situation:** You are analyzing stock performance for a portfolio of 3 companies over 10 trading days. Your investment team wants to know each stock's daily returns, volatility (risk), and total return. This scenario introduces `np.diff()` for calculating day-over-day changes.

### Setting Up the Portfolio Data

We have daily closing prices for three stocks. We store them individually and also combine them into a 2D array for portfolio-level analysis.

```python
np.random.seed(42)  # Reproducible results

# Simulated daily closing prices
days = [f"Day {i+1}" for i in range(10)]
stock_a = np.array([100, 102, 98, 105, 103, 108, 112, 109, 115, 118])
stock_b = np.array([50, 51, 49, 52, 53, 55, 54, 56, 58, 57])
stock_c = np.array([200, 195, 198, 202, 210, 208, 215, 220, 218, 225])

# Combine into a 2D array: rows = days, columns = stocks
portfolio = np.array([stock_a, stock_b, stock_c]).T
print(f"Portfolio shape: {portfolio.shape}")  # (10, 3)
print()

# Display the data
print(f"{'Day':<8} {'Stock A':>10} {'Stock B':>10} {'Stock C':>10}")
print("-" * 40)
for i in range(len(days)):
    print(f"{days[i]:<8} ${stock_a[i]:>8} ${stock_b[i]:>8} ${stock_c[i]:>8}")
```

### Calculating Daily Returns

Daily percentage returns measure how much a stock gained or lost each day, relative to the previous day's price. The formula is:

**Return = (Today's Price - Yesterday's Price) / Yesterday's Price x 100**

We use `np.diff()` to compute the day-over-day price changes, then divide by the previous day's prices. This is a great example of a function that encapsulates a reusable calculation.

```python
def daily_returns(prices):
    """Calculate daily percentage returns from price array."""
    return np.diff(prices) / prices[:-1] * 100

returns_a = daily_returns(stock_a)
returns_b = daily_returns(stock_b)
returns_c = daily_returns(stock_c)

print(f"Stock A returns: {np.round(returns_a, 2)}")
# Stock A returns: [ 2.   -3.92  7.14 -1.9   4.85  3.7  -2.68  5.5   2.61]

print(f"Stock A avg daily return: {np.mean(returns_a):.2f}%")
# Stock A avg daily return: 1.92%
```

### Risk Analysis

Volatility is the standard deviation of daily returns — it measures how unpredictable a stock is. Higher volatility means more risk. We wrap the analysis into a function that returns a dictionary of metrics, combining functions (Week 4), NumPy statistics (Week 5), and dictionaries (Week 3).

```python
def analyze_risk(name, prices):
    """Analyze risk metrics for a stock."""
    returns = daily_returns(prices)
    return {
        "name": name,
        "volatility": np.std(returns),
        "max_gain": np.max(returns),
        "max_loss": np.min(returns),
        "total_return": (prices[-1] - prices[0]) / prices[0] * 100
    }

print(f"{'Stock':<10} {'Volatility':>12} {'Max Gain':>10} {'Max Loss':>10} {'Total Return':>14}")
print("-" * 60)

for stock_name, prices in [("A", stock_a), ("B", stock_b), ("C", stock_c)]:
    metrics = analyze_risk(stock_name, prices)
    print(f"Stock {metrics['name']:<5} "
          f"{metrics['volatility']:>11.2f}% "
          f"{metrics['max_gain']:>9.2f}% "
          f"{metrics['max_loss']:>9.2f}% "
          f"{metrics['total_return']:>13.1f}%")
# Stock A: volatility=3.67%, total return=18.0%
# Stock B: volatility=2.13%, total return=14.0%
# Stock C: volatility=3.42%, total return=12.5%
```

### Interpreting the Results

- **Stock A** has the highest total return (18.0%) but also the highest volatility (3.67%) — higher reward comes with higher risk.
- **Stock B** is the most stable (volatility 2.13%) with a solid 14.0% return — a conservative choice.
- **Stock C** has moderate risk and the lowest total return — the least attractive option in this period.

| Concept | How It Was Used |

## 6.4 Business Scenario 3: Inventory Management

**The situation:** You manage inventory for an electronics store. You need to identify which products need reordering, calculate the restocking costs, and find the most valuable items currently in stock. This scenario showcases boolean indexing, `np.argsort()`, and list comprehensions.

### Setting Up the Inventory Data

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor",
            "Headset", "Webcam", "USB Hub", "Mousepad"]

stock_levels = np.array([5, 150, 45, 12, 8, 25, 60, 200])
reorder_points = np.array([10, 50, 20, 10, 15, 10, 30, 100])
unit_costs = np.array([650, 12, 35, 280, 45, 55, 18, 5])
order_quantities = np.array([20, 200, 100, 15, 50, 50, 100, 500])

# Display inventory status
print(f"{'Product':<12} {'Stock':>8} {'Reorder Pt':>12} {'Unit Cost':>12}")
print("-" * 48)
for i in range(len(products)):
    print(f"{products[i]:<12} {stock_levels[i]:>8} {reorder_points[i]:>12} ${unit_costs[i]:>10,.2f}")
```

### Identifying Items to Reorder

Boolean indexing is the natural tool here: compare stock levels against reorder points to create a True/False mask, then use list comprehension to extract the product names.

```python
# Which items are at or below reorder point?
needs_reorder = stock_levels <= reorder_points
print(f"Reorder mask: {needs_reorder}")
# [ True False False False  True False False False]

# Get the product names that need reordering
reorder_names = [p for p, needed in zip(products, needs_reorder) if needed]
print(f"Reorder needed: {reorder_names}")
# Reorder needed: ['Laptop', 'Headset']
```

### Generating a Reorder Report

This function brings everything together: boolean indexing to identify items, a for loop to process them, vectorized multiplication for cost calculations, and f-strings for a professional report.

```python
def generate_reorder_report(products, stock, reorder_pts,
                            costs, order_qtys):
    """Generate a reorder report with costs."""
    needs_reorder = stock <= reorder_pts
    total_cost = 0

    print("=== REORDER REPORT ===")
    for i, product in enumerate(products):
        if needs_reorder[i]:
            cost = costs[i] * order_qtys[i]
            total_cost += cost
            print(f"  {product}: order {order_qtys[i]} units "
                  f"@ ${costs[i]:,.2f} = ${cost:,.2f}")

    print(f"\nTotal reorder cost: ${total_cost:,.2f}")
    return total_cost

cost = generate_reorder_report(products, stock_levels,
                               reorder_points, unit_costs,
                               order_quantities)
# === REORDER REPORT ===
#   Laptop: order 20 units @ $650.00 = $13,000.00
#   Headset: order 50 units @ $45.00 = $2,250.00
#
# Total reorder cost: $15,250.00
```

### Inventory Value Insights

Use vectorized multiplication to calculate inventory value for each product, then `np.argsort()` to rank them. This shows how NumPy replaces what would otherwise be complex sorting logic.

```python
# Total inventory value
inventory_value = stock_levels * unit_costs
print(f"Total inventory value: ${np.sum(inventory_value):,.2f}")
# Total inventory value: $11,935.00

# Top 3 most valuable items in stock
value_order = np.argsort(inventory_value)[::-1]  # Sort descending
print("\nTop 3 by inventory value:")
for idx in value_order[:3]:
    print(f"  {products[idx]}: ${inventory_value[idx]:,.2f}")
# Top 3 by inventory value:
#   Monitor: $3,360.00
#   Laptop: $3,250.00
#   Mouse: $1,800.00
```

## 6.5 The Business Analytics Workflow

Every scenario we covered followed the same pattern. Recognizing this workflow will help you approach any business analytics problem systematically — whether on the midterm or in your career.

### The 5-Step Pattern

| Step | Description | Python Concepts Used |
|------|-------------|---------------------|
| **1. Define the problem** | What question are we answering? | Understanding the task |
| **2. Organize the data** | Structure data for analysis | NumPy arrays, lists, dictionaries |
| **3. Write functions** | Create reusable logic for calculations | `def`, parameters, `return` |
| **4. Analyze** | Apply calculations and find insights | NumPy stats, boolean filtering, aggregation |
| **5. Present results** | Communicate findings clearly | f-strings, formatted output |

```python
# The workflow in action - a mini example

# Step 1: DEFINE - What's the average order value by customer segment?

# Step 2: ORGANIZE
segments = ["Standard", "Premium", "Enterprise"]
order_values = np.array([
    [50, 75, 60, 45, 80],      # Standard
    [200, 350, 275, 400, 300],  # Premium
    [1200, 1500, 900, 1800, 1100]  # Enterprise
])

# Step 3: WRITE FUNCTIONS
def segment_report(segments, order_values):
    """Summarize average order value by segment."""
    averages = order_values.mean(axis=1)
    for seg, avg in zip(segments, averages):
        print(f"  {seg}: ${avg:,.2f}")
    return averages

# Step 4: ANALYZE
print("Average Order Value by Segment:")
averages = segment_report(segments, order_values)

# Step 5: PRESENT
best_idx = np.argmax(averages)
print(f"\nHighest-value segment: {segments[best_idx]} (${averages[best_idx]:,.2f})")
```

## 6.6 Common Mistakes to Avoid

These are the most frequent errors we see in quizzes, assignments, and exams. Study this table carefully — avoiding these mistakes can save you significant points on the midterm.

| Mistake | What Goes Wrong | How to Fix It |
|---------|----------------|---------------|
| Using `and`/`or` with NumPy arrays | Python's `and`/`or` doesn't work element-wise on arrays | Use `&` (and) and `\|` (or) with parentheses around each condition |
| Forgetting `axis` parameter | `sales.sum()` gives the grand total, not per-row or per-column | Specify `axis=0` (columns) or `axis=1` (rows) |
| Modifying global variables inside functions | Hard to debug, unpredictable side effects | Pass data as parameters, return results |
| No docstrings in functions | Others (and future you) can't understand what the function does | Always add a one-line docstring |
| Not testing edge cases | Code breaks on empty lists, zero values, or unexpected input | Test with `[]`, `0`, `None`, single-element inputs |

```python
# Example: The & vs 'and' mistake

sales = np.array([125, 89, 245, 56, 310, 178, 95])

# WRONG - this will raise an error:
# mid_range = sales[sales > 100 and sales < 200]  # ValueError!

# CORRECT - use & with parentheses:
mid_range = sales[(sales > 100) & (sales < 200)]
print(f"Sales between $100 and $200: {mid_range}")
# Sales between $100 and $200: [125 178]
```

```python
# Example: The axis mistake

data = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(f"No axis (grand total): {data.sum()}")       # 210
print(f"axis=0 (sum down columns): {data.sum(axis=0)}")  # [50 70 90]
print(f"axis=1 (sum across rows): {data.sum(axis=1)}")   # [ 60 150]
```

```python
# Example: Global vs local variables

# BAD - modifying a global variable
total = 0
def bad_add(x):
    global total
    total += x  # Side effect - modifies global state!

# GOOD - pure function with return value
def good_add(values):
    """Calculate the sum of values."""
    return sum(values)

result = good_add([10, 20, 30])
print(f"Sum: {result}")  # Sum: 60
```
